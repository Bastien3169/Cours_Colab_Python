# **0- LES CONDITIONS BOOLEENS**

```python
# Qd le programme voit "pluie", il lit "True" donc continue le programme suivant.
pluie = True
if pluie :
  print("Je prends mon parapluie")
else :
  print("Je ne prends pas mon parapluie")


# Qd le programme voit "pluie", il lit "False" donc passe au programme suivant.
pluie = False
if pluie :
  print("Je prends mon parapluie")
else :
  print("Je ne prends pas mon parapluie")


# Exemple pour arrêter une chaine en continuant une itération : mettre une seule majuscule
phrase = 'un Denier Exercice Et Vous Pourrez Vous Reposer.'
majuscule = True
np = ""

for i in phrase:
    if majuscule :
        i=i.upper()
        majuscule = False
    else:
        i=i.lower()
    np = np + i
print(np)


# Exemple pour mettre un caractere sur deux en majuscule
phrase = 'un Denier Exercice Et Vous Pourrez Vous Reposer.'
majuscule = True
np = ""

for i in phrase:
    if majuscule :
        i=i.upper()
        majuscule = False
    else:
        i=i.lower()
        majuscule = True
    np = np + i
print(np)
```

# **I- LES CONDITIONS IF (si), ELIF(sinon si) ET ELSE (sinon)**

```python
#Déclaration variable
bitcoin_price = 45000

#SI Bitcoin <= 3000 on print "Acheter"
if bitcoin_price <= 30000:
    print("Acheter")
#SINON SI Bitcoin est > à 3000 on print "Vendre la moitié"
elif bitcoin_price > 30000 and bitcoin_price <= 50000:
    print("Vendre la moitié")
#SINON SI Bitcoin est < à 60000 on print "Conserver"
elif bitcoin_price < 60000:
    print("Conserver")
#SINON (si rien de tout ça), on print "Vendre la totlaité"
else:
    print("Vendre la totalité")
```

# **II- BOUCLE FOR (pour...)**

```python
liste = [1, 2, 3]

# Effectuer une opération pour chaque element d'une "liste", "dictionnaire" ou même "chaine de caractère" ou fonction "range()"

#La boucle for parcoure (ou itère) chaque élément de la liste.
#À chaque itération de la boucle, la variable "element" prend la valeur de l'élément actuel de la liste.
for element in liste:
#Ce "print" imprime la valeur "element" à chaque itération de la boucle.
    print(element)
#Ce "print" imprime le dernier élément itéré
print(element)
```

```python
# Créer une liste à partir d'une liste vide avec la boucle "for":
variable = "Bonjour, je fais un test"
liste=variable.split(" ")

test = []
for i in liste:
  total=i*2
  test.append(total)
print(test)


#OU 2eme solution "compréhention de liste" est une syntaxe concise et expressive qui permet de créer des listes d'une manière élégante.
test = [i * 2 for i in liste]#En faite on met [], on voit qu'on est dans une liste et on fait le calcul et l'itération directment dedans.
print(test)
```

```python
liste1 = [1, 6, 3]
liste2 = [4, 5, 6]


# multiple itération boucle "for" :
for a, b in zip(liste1, liste2):
  calcul = a*b
  print(calcul)


# Condition avec plusiuers itération .zip(): Attention, mettre "all()" sinon ça renvoie "true" car En Python, un générateur est un objet et lorsqu'il est évalué dans une déclaration "if", il est toujours considéré comme True, car l'existence de l'objet générateur lui-même est vérifiée, pas le contenu du générateur.
if all(x<y for x,y in zip(liste1,liste2)):
  print("true")
else:
  print("false")


# boucle impriquée "for" :
print("----- Ex1: boucle imbriquée-----")
for i in liste1:
  test1 = i+1
  print(test1)
  for b in liste2:
    test2 = b+2
    print("______")
    print(test2)

print("----- Ex2: boucle imbriquée-----")
n = range(1,13)
for i in n:
    for y in n :
        print(f"{i} x {y} = {i*(y)}")
    print("---")

print("----- Ex3: boucle imbriquée-----")
joueurs = ["Alice", "Bob", "Carole"]
for i in joueurs:
    for y in joueurs:
        if i == y:
            pass
        else:
            print(f"{i} contre {y}")
```

#**III- BOUCLE WHILE (tant que...)**

```python
#Déclaration des variable
invest_ini = 1000
obj_cap = 2000
taux = 0.04

#Initialisation nombre d'année
année=0

#Boucle pour atteinde l'objectif de 2000€ de capital
while invest_ini < obj_cap:
#alors, on incrémente  le taux à mon investissemennt ini juqu'à avoir 2000€
  invest_ini = invest_ini + (invest_ini * taux)
#Combien d'année ça va prendre ?
  année = année + 1

print(année)
```

# **IV- EXEMPLE BOUCLE WHILE AVEC CONDITION IF**

```python
#Déclaration des variables
goal=10000
epargne=400
promotion=6
epargne_promotion=200

#Initialisation compteurs (si on avait eu une épargne de 200€, mettre 200€)
mois_necessaire = 0
epargne_actuelle = 0

#Boucle pour atteindre l'objectif d'épargne :
#Tant que mon épargne actuelle est < au goal...
while epargne_actuelle < goal:
#... (vérif si promotion à eu lieu) et si le nbr de mois nécessaire...
#...est = au mois de la promotion...
  if mois_necessaire == promotion:
#... Alors, on ajoute à mon épargne ma promotion de 200€/mois
#Au départ c'est faux donc on rentre pas dedans, d'ou l'indentation.
    epargne = epargne + epargne_promotion

#Mon épargne actuelle est = à mon épargne actuelle incrémentée de...
#... mon épargne pour chaque mois jusqu'à arriver au goal.
  epargne_actuelle = epargne_actuelle + epargne
#Incrémenter (rajouter) chaque fois 1 mois jusqu'au goal
  mois_necessaire = mois_necessaire + 1

print("épargne = ", epargne_actuelle, "/ mois nécéssaire = ",mois_necessaire)
```

# **V- EXEMPLE BOUCLE WHILE ET BOUCLE FOR**

```python
#Déclaration des variable
invest_ini = 1000
obj_cap = 2000

#Création boucle "for" pour incrémenter de 2 à 9 avec un pas de 1 (soit 7 fois).
for taux in range(2,9,1):
#Ici on va donc faire 3 incrémentations:
#- On itère le taux dans le range en divisant par 100 pour la suite du calcul
  taux_en_pourcentage=taux/100
#- On itère invest-ini dans le range en changant de variable
  invest = invest_ini
#- On itère les années dans le range en partant de 0 ans
  annee=0
#Création de la boucle while: TANT QUE invest est < à l'objectif (2000€)...
  while invest < obj_cap:
#...ALORS, on incrémente invest pour chauqe itération juqu'à avoir 2000€
    invest = invest + (invest * taux_en_pourcentage)
#...ALORS, on incrémente le nbr d'annee pour chaque itération juqu'à avoir 2000€
    annee+=1
#Attention à l'indentation pour le print!
  print(taux, invest, annee)
```

# **V- "BREAK", "CONTINU" ET "PASS"**

```python
# 1) break :
'''L'instruction break est utilisée pour sortir d'une boucle de manière anticipée.
Lorsque Python rencontre l'instruction break à l'intérieur d'une boucle (comme for ou while),
il arrête immédiatement l'exécution de la boucle et passe à l'instruction suivant immédiatement après la boucle.
C'est utile lorsqu'une condition spécifique est remplie et que vous souhaitez arrêter l'exécution de la boucle prématurément.'''

for i in range(10):
    if i == 3:
        break
    print(i)
print("-"*10)

# 2) continu :
'''L'instruction continue est utilisée pour passer à l'itération suivante d'une boucle.
Lorsque Python rencontre l'instruction continue à l'intérieur d'une boucle, il arrête
l'exécution de l'itération actuelle et passe à l'itération suivante de la boucle.
C'est utile lorsque vous souhaitez ignorer certaines conditions dans une boucle mais continuer à itérer.'''

for i in range(5):
    if i == 2:
        continue
    print(i)
print("-"*10)

# 3) pass :
'''L'instruction pass ne fait rien.
Elle est souvent utilisée comme un espace réservé lorsque la syntaxe l'exige mais que vous ne souhaitez pas exécuter de code.
Elle est utilisée pour éviter les erreurs de syntaxe lorsque Python attend un bloc de code, mais aucun code n'est nécessaire.'''

for i in range(5):
    if i == 2:
        pass
    else:
        print(i)

```
