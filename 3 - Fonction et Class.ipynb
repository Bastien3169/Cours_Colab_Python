#**I- LES FONCTIONS**

```python
'''Une fonction est un bloc de code auquel on peut donner le nom que l’on veut. Il sert à alléger le code et simplifier la lecture.

#Attention, si on fait un "print" d'une fonction et quelle ne retourne rien, ça fera un "none"!'''

#Il existe deux types de fonctions :

# I-	Les fonctions intégrées (à python) comme :
print(), input(), #etc…

# II-	Les fonctions définies (par le développeur) avec “def“:

  #	1) La fonction sans paramètre ou sans argument (rien dans les “()“)
def interet_compose():
  return interet_compose

  #	2) La fonction avec paramètre
def interet_compose(apport=10000, taux=0.08, temps=10):
    quantite = apport * (1 + taux) ** temps
    interet = quantite - apport
    return interet
print(interet_compose())

  #Si on veut changer les arguments par défauts, on en met de nouveaux en argument:
print(f"L'intérêt composé après {temps} années est de {round(interet_compose(3000,0.04,5), 2)} €")

  #	3) La fonction avec une valeur retour des paramètres
    #Pas sur que ça existe

print(interet_compose()) #Reprend ses argumants par défauts
print(round(interet_compose(), 2)) #La fonction "round()" permet d'arondir, ici, à 2 chiffres.
```

```python
'''
- Le "return" permet de stocker le résultat dans notre fonction "def".
- Ce n’est pas obligatoire de retourner une valeur si l’on ne souhaite pas récupérer de résultat
en particulier (par exemple si on cherche juste à afficher quelque chose)'''

# 1) Sans "return" :

def racine(n):
  r = n * n
  print(r)

try:
  test = racine(5) * 2
  print(test)
except :
  print("Ca print le 5*5 mais pas le (5*5)*2 car pas de rerturn et on peut pas se servir de la racine en tant que int")


# 2) Avec "return" :
print("-"*10)####

def racine(n):
  r = n * n
  return r

test = racine(5) * 2
print(test)
```

```python
#"__init__" = constructeur d'une classe
#"self" représente l'instance de la class. Grace à "self" qu'on accède aux attributs et aux méthodes d'une class.

'''
1) Les attributs:  sont ses caractéristiques (couleur, marque, puissance). Pas besoin de ().
C'est des des variables associées à un objet.


2) Les méthodes: sont ses actions (accélérer, freiner, klaxonner). Il faut des ().
C'est des fonctions définies dans une classe.
'''
```

```python
# I - "class" SANS ARGUMENT fixe et sans constructeur :

import random

class De():

  def lance(self):
    de1 = random.randint(1,6)
    de2 = random.randint(1,6)
    return de1, de2

essaie1 = De()
print(essaie1.lance())
```

```python
# II - "class" AVEC ARGUMENT

# On créé la "class" avec une majuscule (par convention) :
class Snow:
  # Variables de class : communes à toutes les instances. Elles sont fixes et on ne peut pas les changer.
  carre = 1
  fix = 2
  style = "Le snow, c'est cool"

  # Méthode "__init__" (qui est une fonction) : sert à créer (ou initialiser) les variables d'instances.
  def __init__(self, taille, terrain, prix):
    # Variables d'instance : variable pour chaque instance.
    self.taille = taille
    self.terrain = terrain
    self.prix = prix

  # Méthode de class (qui est une fonction). Ici, la méthode à 2 arguments : "self" obligatoire et "couleur" optionnel qui ne servira que pour cette méthode
  def carac_board(self, couleur):
    print("Si tu as un snow de taille de taille", self.taille, "fais du", self.terrain, "et le prix sera", self.prix, "avec une couleur", couleur)


# Création de 3 objets (ou instances) : héritera des variables de "class" qui sont fixes et des variables "d'instances" qu'on leur met pour chaque instance.
snow_freestyle = Snow(taille = "< 155 cm", terrain = "Freestyle", prix = "assez reuch")
snow_backcountry = Snow(taille = "entre 155 cm et 165 cm", terrain = "Backcountry", prix = "ça va, moins reuch")
snow_race = Snow(taille = "> 165 cm", terrain = "piste", prix = "pas très reuch")

# On appelle notre instance avec avec une méthode créée:
snow_freestyle.carac_board("rouge")
print(snow_freestyle.carre)
```

```python
# III - "class" AUTRE EXEMPLE AVEC ARGUMENT

class Personne:
  def __init__ (self, nom = " ", age = 0):
    self.nom = nom
    self.age = age

  def SePresenter (self):
    if self.age == 0 :
      age1=int(input("Entrez votre age: "))
      self.age = age1
    if self.nom == " " :
      nom1=str(input("Entrez votre nom: "))
      self.nom = nom1

    print(f"Bonjour, je m'appelle {self.nom} et j'ai {self.age} ans.")

    if self.age > 18 :
      print("Je suis majeur")
    else:
      print("Je ne suis mineur")

  def EstMajeur (self):
    if self.age > 18:
      return True
    else :
      return False

personne1=Personne(nom = "Robert", age = 33)
personne2=Personne(nom = "Bastien", age = 17)
personne3=Personne()
personne1.SePresenter()
personne2.SePresenter()
personne3.SePresenter()
```

```python
# IV - "class" AUTRE EXEMPLE AVEC ARGUMENT
class DataProfile:
    def __init__(self, name, skills):
        self.name = name
        self.skills = skills

    def display_skills(self):
        print("----------")
        print(f"Compétences de {self.name}:")
        for skill, level in self.skills.items():
            print(f"{skill}: {level}/10")

class DataEngineer(DataProfile):
    pass

class DataScientist(DataProfile):
    pass

class DataAnalyst(DataProfile):
    pass

# Création des profils avec des compétences fictives
engineer = DataEngineer("Data Engineer", {"Hadoop": 7, "Spark": 8, "Kafka": 6, "ETL": 9, "SQL": 8})
scientist = DataScientist("Data Scientist", {"Machine Learning": 9, "Statistiques": 8, "Deep Learning": 7, "Python": 9})
analyst = DataAnalyst("Data Analyst", {"Analyse de Données": 8, "SQL": 9, "Tableau": 9, "Excel": 9})

# Affichage des compétences
engineer.display_skills()
scientist.display_skills()
analyst.display_skills()
```

```python
# IV - "class" AUTRE EXEMPLE AVEC ARGUMENT
class Voiture:
    def __init__(self, marque, modele, couleur = "On ne sait pas", km = "On ne sait pas"):
        self.marque = marque
        self.modele = modele
        self.couleur = couleur
        self.km = km

    def changer_couleur(self, nouvelle_couleur):
        self.couleur = nouvelle_couleur

    def changer_km(self, nouveau_km):
        self.km = nouveau_km

    def __str__(self):
        return f"Voiture: {self.marque} {self.modele}, couleur : {self.couleur}, km : {self.km}"

voiture1 = Voiture("Peugeot", "208")
voiture2 = Voiture("Renault", "Clio")

print(voiture1)
print(voiture2)

voiture1.changer_couleur("Bleu")
voiture1.changer_km("1000 km")

voiture2.changer_couleur("Rose")
voiture2.changer_km("234 km")

print(voiture1)
print(voiture2)
```

#**III- CLASSMETHOD, STATICMETHOD ET DATACLASS**

```python
# 1) Méthode de "class" : @classmethod :
  #Il s'agit d'un décorateur :

# 2) Static de "class" : @staticmethod :
  #Il s'agit d'un décorateur :

# 3) @dataclass : permet de simplifier à la place de la méthode "__init__":

  # Ceci est équivalent à la class "__init__" soit l'exemple en dessous :
from dataclasses import dataclass

@dataclass
class Voiture :
  couleur = str
  prix = str

  # Ceci est équivalent à "@dataclass" soit l'exemple au dessus :
class Voiture :
  def __init__(self, couleur, prix):
    self.couleur = couleur
    self.prix = prix
```

#**IV- L'ENCAPSULATION, L'HERITAGE, FONCTION '__dict__'**

```python
# 1) L'ENCAPSULATION : attribut privé : __attribut : empecher l'accés aux données par un autre moyen que définit.
class Animal:
    def __init__(self, race):
        self.__race = race #Attribut privé : ne peut pas être modifié ou écrasé

    def autre_race(self):
        return self.__race

# 2 moyens de parvenir à cet attribut privé :
le_lion = Animal("Lion")
print(le_lion.autre_race()) # .get : appeler l'attribut privé

le_singe = Animal("Singe")
print(le_singe._Animal__race) # ._class__attribut privé : ._Animal__race


# 2) HERITAGE d'une classe :
class Felin():
  nbr_pattes = 4
  carnivor = True

  def __init__ (self, couleur, race):
    self.couleur = couleur
    self.race = race

  def manger_viande():
    print("La viande c'est bon")

class Lion(Felin): # La calsse Lion devient class fils de Felin
  def __init__ (self, couleur, race):
    super().__init__(couleur, race)

lion_savane = Lion("noir", "savane")
print(lion_savane) #Ca marche pas ?? ça print un objet.


# 3) Fonction __dict__ : permet de récuperer dnas une dictionnaire arguments de chaque méthodes (variables) des instances.
lion_savane.__dict__


# 4) Unpacker des arguments : "*args"
  # Permet de récuprer un nombre infini d'argument
```
