#**III- ACCEDER AUX LIGNES (dataframe) : .loc, .iloc et slice Python**

```python
# 1) ACCEDER AUX LIGNES AVEC LES METHODES .loc[], .iloc[] et [] :

# .index() : Acceder au étiquette de l'index
print("______ .index()______")
display(df2.index) # Récupère info index (start=0, stop=5, step=1)

# .loc[] : Méthode recommandée
print("______ .loc[] Recommandée______")
display(df2.loc[2:2])

'''
- Sélectionne en utilisant les étiquettes (labels en anglais) d'index (indexation basée sur les étiquettes) : Méthode recommandée
- Syntaxe : df.loc[row_label, column_label].
- Inclue la fin de l'intervalle
- Plus explicite et recommandée lorsque l'index du DataFrame est basé sur des étiquettes.

Exemple :
# df.loc[2:9] : récupère les lignes de la colonne index 2 à 9 (9 étant inclu !).
# df.loc["ezf@gmail.com"] : On récupère tous head de chaque colonne avec leursdonnées de cet index email.
# df.loc["ezf@gmail.com"].values : Mets uniquement les données de chaque colonnes (sans les titres des colonnes) sous forme de liste spécial objet python.
# df.loc["ezf@gmail.com"].values.tolist() : Mes sous forme de liste
# df.loc[["ezf@gmail.com","fdds@gmail.com"]] : récupère les lignes de la colonne index 2 à 9 (9 étant inclu !).
'''

# .iloc[] "i" pour "indice": Equivalent à un "slice" : Méthode peu fiable
print("______ .iloc[] Peu fiable______")
display(df2.iloc[1:3])

'''
- Selectionne en utilisant les indices (indexation basée sur la position). Méthode peu fiable
- Les arguments sont des indices de position.
- Syntaxe : df.iloc[row_index, column_index].
- Exclut la fin de l'intervalle.
- Utile lorsque l'index du DataFrame est un index entier (0, 1, 2, ...).

Exemple :
# df.iloc[2:9] : récupère les lignes de la colonne index 2 à 9 (9 étant exclu !).
'''

# [] : Equivalent .iloc[] et donc à un "slice"
print("______ .[] idem .iloc[]______")
display(df2[1:3])

'''
Même chose que df.iloc[]
'''
```

#**IV- ACCEDER A UNE VALEUR PRECISE (élément unique) : .loc; .iloc; .at; .iat**

```python
'''
Utilisez .loc et .iloc pour accéder ou modifier des sous-ensembles de données ou lorsque vous avez besoin de la flexibilité qu'ils offrent.
Utilisez .at et .iat pour accéder ou modifier un seul élément et que la performance est cruciale.

- .loc et .iloc : moins optimisé car même si une valeur, cherche en plusieurs étapes.
- .at et .iat : optimisation car cherche qu'une valeur donc q'une étape.


'''
# .loc() : par label de colonne : df2.loc[label index, label colonne]
print("______ df2.loc[label ligne, label colonne]______")
display(df2.loc[2,"Test"])


# .iloc() : par indice de colonne "i" pour "indice" : df2.iloc[indice ligne, indice colonne]
print("______ df2.iloc[indice ligne, indice colonne]______")
display(df2.iloc[2,2])


# .at() : par label de colonne : df2.at[label ligne, label colonne]
print("______ df2.at[indice ligne, label colonne]______")
display(df2.at[2,"Test"])


# .iat() : par indice de colonne "i" pour "indice" : df2.iloc[indeci ligne, indice colonne]
print("______ df2.iat[indice ligne, indice colonne]______")
display(df2.iat[2,2])


# 3) slice[] : moins optimisé.
'''
2 défauts :
- Plus lent car 2 étapes : fait une séléction de colonne puis un slice sur cette selection.
- Moins flexible car ne fonctionne pas avec des intervales en str ou des selections en ligne car cherche colonne.
'''
print("______ slice[] avec selection colonne précise + intervale ligne (slice) : df2['Pays'][0:2]______")
display(df2["Test"][2])
```

```python
# 1) .loc[] : par label index et colonne : df2.iloc[label index, label colonne]

  # 1.1) Accéder à une plage (slice) de lignes et de colonnes sélèctionnées:
'''
Sélection d'une plage de colonnes et des lignes en une étape. Si une colonne entre, on l'aura ici il y a 'Test':
- df2.loc[0:2, "Pays":"Taille"] utilise .loc pour sélectionner directement les lignes de 0 à 2
et les colonnes de 'Pays' à 'Taille' (inclusivement, si elles sont dans cet ordre dans le DataFrame).
'''
print("______ df2.loc[0:2, 'Pays':'Taille']______")
print(df2.loc[0:2, "Pays":"Taille"])

  # 1.2) Accéder à des sélections de lignes et/ou colonnes dans une liste :
print("______ df2.loc[0:2, 'Pays':'Taille']______")
print(df2.loc[[0,2], ["Pays","Taille"]])

  # 1.3) Selection + Intervalle et inversement : On peut évidemment mixer selection et intervalle
print("______ df2.loc[0:2, 'Pays':'Taille']______")
print(df2.loc[0:2, ["Pays","Taille"]])


# 2) .iloc[] : par indice index et colonne "i" pour "indice" : df2.iloc[indice ligne, indice colonne]
print("______ print df2['Pays'].iloc[1]______")
print(df2["Pays"].iloc[0:2])


# 3) slice[] : moins optimisé.
'''
2 défauts :
- Plus lent car 2 étapes : fait une séléction de colonne puis un slice sur cette selection.
- Moins flexible car ne fonctionne pas avec des intervales en str ou des selections en ligne car cherche colonne.
'''
print("______ slice[] avec selection colonne précise + intervale ligne (slice) : df2['Pays'][0:2]______")
print(df2["Pays"][0:2])

print("______ slice[] avec selection plusiseurs colonnes + intervale ligne (slice) : df2[['Pays','Taille']][0:2]______")
print(df2[["Pays", "Taille"]][0:2])

#print(df2["Pays":"Taille"][0:2]) fonctionne pas à cause ["Pays":"Taille"] (Python ne gère pas les slices en str, il faut passer par pandas et donc .loc ou .iloc)
#print(df2[["Pays", "Taille"]][[0,2]]) fonctionne pas à cause ["0","2"] (cherche à reselectionner colonne mais trouve pas "0" et "2" car n'existe pas)


# 4) slice[] + .loc[] : moins optimisé car utilise 2 méthodes (slice python + .loc)
'''
Deux étapes distinctes, car 2 méthodes :
  - Slice de Python sélectionne la/les colonnes.
  - .loc (ou.iloc) de Pandas, sélection la/les lignes.
'''
  # Acceder à des intervalles (slice) de lignes :
print("______ print df2['Pays'].loc[0:3]______")
print(df2["Pays"].loc[0:2])

  # Accéder à des sélections de lignes et/ou colonnes dans une liste :
print("______ df2[['Pays','Taille']].loc[0:2]______")
print(df2[["Pays","Taille"]].loc[[0,2]])


# 5) Acceder à des lignes sélèctionnés de colonnes avec .head(), .tail(), .sample(), etc...
print("______ df2.['Pays'].head(2)______")
print(df2["Pays"].head(2))
```

#**VII- MODIFICATION D'UN TABLEAU AVEC .rename; .replace; .loc; .iloc; .at; .iat**

```python
'''
Lorsqu'on modifier un dataframe via sa variable "df" par exemple, ça ne la modifie pas dans la mémoire donc n'enregistre pas les midifications.

Pour cela, 3 méthodes :

- df_copie = df.copy().drop("gender", axis = 1) # On fait une copie de la dataframe dans une nouvelle variable et on fait nos modifs
- df.drop("gender", axis = 1, inplace=True) # On utilise l'argument "inplace" et on met sur "True" (car "False par défaut) et ça écrase notre dataframe
- df = df.drop("gender", axis = 1) # On ecrase notre dataframe avec une nouvelle variable identique.

À noter que la méthode .rename(), .replace(), ne modifie pas le data frame existant. Il existe cependant
un argument pour cette méthode (et pour toutes les méthodes similaires) nommé  inplace , qu’il suffit
de fixer à Vrai (True ) pour pallier cela. Ainsi,  data.rename(columns={'identifiant': 'ide'}, inplace=True)  est
strictement équivalent à  data = data.rename(columns={'identifiant': 'ide'})

'''

# 1) REMPLACER/MODIFIER
'''
!!!! .str.replace() (ou rename ou contains) fonctionne pour objet DataFrame et objet Serie.
A l'inverse .replace n'existe que pour des objet Serie !!!
'''

# .rename() : Rename étiquette "columns" ou "index" avec un dico :
print("______ .rename(columns = {'Pays':'Country'}) ______")
display(df2.rename(columns={'Pays':'Country'})) # Rename une colonne avec un dico. clé : valeur à rename / Valeur : valeur rename
print("______ .rename(index = {2:'index_2'}) ______")
display(df2.rename(index={2:"index_2"})) # Rename un index avec un dico. clé : valeur à rename / Valeur : valeur rename
print("______ .rename(index et columns) ______")
display(df2.rename(index={2:"index_31"}, columns={'Pays':'zeubi'}))

# .replace() : Replace valeur ds tout le DataFrame avec un dico (idem .rename() mais pour les valeurs:)
print("______ .replace() int vers str, .replace({10:'tutut'}) ______")
display(df2["Poids"].replace({10:"tutut"})) # Change le int 10 par le str "io"

print("______ .replace() une valeur, .replace({'Chien':'Toulouse'}) ______")
display(df2["Animal"].replace({"Chien":"Toulouse"}))

print("______ .replace() plusieurs valeurs .replace({'Chien' : 'best', 'Chat' : 'bad'}) ______")
display(df2["Animal"].replace({"Chien" : "best", "Chat" : "bad"})) # Replace plusieurs valeurs avec un dico. clé : valeur à replace / Valeur : valeur replace

print("______ .replace() une lettre, .replace({'e':'3'}) ______")
display(df2["Animal"].str.replace("e","3")) # .str car objet serie pandas et doit appliquer la méthode sur chaque élément de la série.
display(df2.loc[3,"Animal"].replace("e","3")) # pas de .str car l'objet est une chaîne de caractères.


# .loc[] (par label index et colonne) et .iloc[] (par indice index et colonne) :
''' Les exemples seront avec .loc[] mais fontionnent aussi avec .iloc[]'''

  # Pour remplacer une valeur précise
print("______ .loc[] : df2.loc[2, 'Pays'] = 'Ski' ______")
modif = df2.loc[2, "Pays"] = "Ski" # Remplace une valeur précise
display(df2)

  # Pour remplacer une selection de valeur
print("______ .loc[] : df2.loc[1:4,'Pays'] = 'Ski' ______")
modif = df2.loc[1:4,"Pays"] = "Ski"
display(df2)

 # Avec condition : "mon_dataframe.loc[ condition sur les lignes, colonne(s)]""
print("______ .loc[] : df2.loc[df2['Animal']=='Chien', df2['Poids']>10] = 281727 ______")
modif = df2.loc[df2['Animal']=="Chien", "Poids"] = 281727 # Si mon "animal" est un "chien" alors change le poids à 281727
display(modif)

# .at[] et .iat[] : Pour remplacer uniquement une seul valeur précise mais plus optimisé que .loc ou .iloc.
print("______ .at[] : df2.at[1, 'Pays'] = 'Snowboard' ______")
modif = df2.at[1, "Pays"] = "Snowboard"
display(df2)
```

```python
# .set_index : Remplace l'indexe par défaut par une colonne du tableau
print("______ .set_index('Pays')______")
print(df2.set_index("Pays")) # Remplace l'indexe numérique par l'indexe colonne "Pays"

# .reset_index : Revenir à un index numérique par défaut (on peut mettre : inplace=True)
print("______ .reset_index ______")
df2.reset_index(drop=True)#Supprime l'ancien index au lieu de l'ajouter comme colonne au DataFrame.
display(df2)

# .set_index() : modifie la colonne index par une étiquette de colonne existante dans le tableau, ici, l'index n'est plus les chiffres mais les "Pays"
print("______ .set_index('Pays')______")
print(df2.set_index("Pays")) # modifie la colonne index par une étiquette de colonne existante dans le tableau, ici, l'index n'est plus les chiffres mais les "Pays"

# .split() : split que sur du str
print("______ .split()______")
'''
Ne fonctionne qu'avec .str car que sur chaine de caractère.
L'argument "expand = True" étend les résultats du split en plusieurs colonnes.
'''
df2["New col split"] = df2["Pays"].str.split("/") # Sans 'expand': met la liste du 'split' dans la colone
display(df2)
df2["New col split"] = df2["Pays"].str.split("/").str[0] # Sans 'expand' et [0]: met dans la colonne l'élement [0] de la liste du split
display(df2)
df2[["New col split1", "New col split2"]] = df2["Pays"].str.split("/", expand=True) # Avec 'expend' : met chaque éleent de la liste dans une nouvelle colonne
display(df2)

# .get() : récupère l'indice '0' après avoir splitté sur '/'
print("______ .get(0) : récupère l'indice '0' après avoir splitté sur '/'______ ")
df3["Pays_cool"] = df2["Pays"].str.split("/").str.get(0) # Créé une nouvelle colonne avec ce qi'il y a en indice 0 de la colonne "Pays" après l'avoir splitté
display(df3)

# .upper() et .lower() : pas besoin de faire un dessein !

# .astpyte() : modification des types d'objets
print("______ .astype()______")
df2["Poids"] = df2["Poids"].astype(str) # Change le type "float" en "str"
display(df2.dtypes)

# pd.to_numeric : modification des types d'objets d'une série pandas en valeurs numériques.
'''
Argument errors :
 - errors='raise' (par défaut) : Lève une exception si une conversion échoue.
 - errors='coerce' : Remplace les valeurs non convertibles par NaN (Not a Number). Cela est très utile pour identifier et gérer les valeurs incorrectes sans interrompre le processus de conversion.
 - errors='ignore' : Retourne les données originales sans faire de conversion si une conversion échoue.
'''
df2['Test3'] = pd.to_numeric(df2['Test2'], errors='coerce')
display(df2)

'''
Il est possible de faire des opérations du même type sur 2 classeurs ou 2 feuilles de calcul.
'''
```

```python
# 1-1) CONCATENATION VERTICALE (PAR DEFAUT) AVEC "pandas.concat()"
'''
Assemblage de plusieurs DF ayant la même structure, mêmes colonnes, mêmes types, mais des informations différentes.
Les DF sont mis bout à bout, l’un à la suite des autres (comme lors d’un ajout d'élément à une liste).

La concat ne gère les index par défaut e recommance à 0 sur le 2eme DF qui a été concaténé : index en doublon!!
Pour l'éviter : utiliser l’argument "ignore_index=True" , qui équivaut à appliquer une méthode "reset_index" :

#Par défaut :
pd.concat([df1, df2], axis=0, ignore_index=False, join='outer')

# Il faut donc faire
pd.concat([df1, df2], ignore_index=True) .
'''

# Par défaut, "ignore_index=False" et "axis=0": les indexs reprennent (donc doublons) et cancat verticale
print("______ concat sans ingore index : print(pd.concat([df2, df3])) ______")
display(pd.concat([df2, df3]))

# "ignore_index=True" : les indexs continuent et pas de doublon
print("______ concat sans ingore index : print(pd.concat([df2, df3])) ______")
display(pd.concat([df2, df3], ignore_index=True))

# Filtratrion du DF de la même façn ensuite :
print("______ filtre ['Poids'] > 6 sur la colonne  ______")
tab_concat = pd.concat([df2, df3], ignore_index=True) # On met dans une variable
tab_conact_sup6 = tab_concat["Poids"] > 6 # On filtre la colonne "Poids" sur > 6
display(tab_conact_sup6[tab_conact_sup6]) # On print le filtre sur la colonne


# 1-2) CONCATENATION HORIZONTALE AVEC "pandas.concat()"

# Par défaut, "axis=0" : concaténation verticale donc si concat horizontale "axis=1"
print("______ concat horizontal : print(pd.concat([df2, df3], axis=0)) ______")
display(pd.concat([df2, df3], axis=0))

# Jointure sur colonne précise en les méttant à la place de l'index avec "set_index()":
'''
Les valeurs de la colonne bouge et passe en index et n'existe plus en colonne. Si "set_index" différent
pour chasue DF, il y aura des "NaN" sur les valeurs déplacées en index.
'''
print("______ jointure sur colonne 'Age' en index : print([df2.set_index('Age'), df3.set_index('Age')], axis=0)) ______")
display(pd.concat([df2.set_index("Age"), df3.set_index("Age")], axis=0))
print("______ jointure sur colonne 'Age' et 'Test' en index : print([df2.set_index('Age'), df3.set_index('Age')], axis=0)) ______")
display(pd.concat([df2.set_index("Age"), df3.set_index("Test")], axis=0))


# 1-3) AUTRES :

# Avec méthode ".append()", mais ne va plus exister ds prochaine version
'''display(df2.append(df3))'''

#A ne pas confondre avec de la concat de print qui ajoute en horizontal
print("______ print(df2 + df3) ______")
display(df2 + df3) # Concac pandas mauvais pratique
```

```python
# 3) FUSIONNER (MERGE) AVEC : pandas.DataFrame.merge()
'''
Par défaut :
pd.merge([df1, df2, df3], axis=0, ignore_index=False, join='outer')

Fusionne les colonnes communes en ligne avec un index avec jointure des cellules vides (inverse pour join='inner')
'''

'''
2 méthodes identiques :
- Fonction Pandas : pd.merge(A, B)
- Méthode DataFrame : A.merge(B)
'''

'''
On part du DF de gauche, on reste sur celui de gauche et on merge sur lui le DF de droite en indiquant nos identifiants.

# jointure entre 2 dataframes A et C
C.merge(A, left_on='identifiant', right_on='id')

# jointure entre 2 dataframes A et B
pd.merge(A, B, left_on='id', right_on='identifiant', how = 'inner)

4 types de jointures :
- interne (inner ou jointure naturelle (par défauts si pas de clé)) : conserve que clés identiques aux 2 DF.
- gauche (left) : Concerve que clés du DFgauche et complète ces mêmes clés avec celle du DFdroite. Si valeur n'existent pas sur DF de droite: ça rajoute dans la case un "NaN".
- droite (right) : Inverse de la jointure de gauche
- externe (outer) : Conserve TOUTES les clés du DFgauche et DFdroite. Bouche les info manquantes par de "NaN".

- indicator=True : ajoute une colonne qui indique d'ou viennent les valeurs (right_only, left_only ou both)
'''
print("______ (pd.merge(df2, df3, on='Animal', how='inner')) ______")
display(pd.merge(df2, df3, on="Animal", how='inner'))
print("______Avec 'indicator': (pd.merge(df2, df3, on='Animal', how='inner', indicator=True) ______")
display(pd.merge(df2, df3, on="Animal", how='inner', indicator=True))


```
