#**I- LES ERREURES AVEC "try/except"**

```python
# Exemple division d'erreur python: ZeroDivisionError, NameError, ValueError, ou Typeerror
#try except finally

  # Exemple 1 division zero:
def division(nominateur, denominateur):
    try:
        resultat = nominateur / denominateur
        print(resultat)
    except ZeroDivisionError :
        print("Une division ne peut pas être divisé par 0 ")

division(2,0)


  # Exemple 2 plusieurs erreurs:
def division(nominateur, denominateur):
    try:
        resultat = nominateur / denominateur
        print(resultat)
    except ZeroDivisionError:
        print("Erreur : Une division ne peut pas être effectuée par 0.")
    except TypeError:
        print("Erreur : Les deux valeurs doivent être des nombres.")
    except Exception as e:
        print(f"Erreur inattendue : {e}")


division(2, 0)
division(2, "a")
division(2, 2)

# Créer une erreure : avec "raise" :
a = 5
b =2

if b < 5 :
  print("'b' est inférireur à 5")
  raise ValueError
else :
  c = b * a

print(c)
```

#**II- ReGex : expressions régulières**

```python
# Sert à retrouver certaines chose
#Module re

'''Les tokens communs :

[abc] : Un caractère unique de a,b ou c
[^abc]: Un caractère sauf a,b ou c
[a-z] : Un caractère entre a et z
[^a-z] : Un caractère pas entre a et z
[a-zA-Z] : Un caractère entre a-z ou A-Z
. : N'importre quel caractère
\s : n'importe espace blanc caractère
\S : N'importe non espace blanc caractère
\d : N'importe quel nombre
\D : N'importe caractère qui ne soit pas un nombre
\w : N'importe mot
\W : N'importe non mot
(...) Récupérer tout ce qui est entre parenthèse
(a|b) : Match a our b
a* 0 : ou + occurences de a
a+ 1 : ou + occurences de a
a{3} : Exactement 3 occurences de a
a{3,}: 3 ou + occurence de a
a{3,6} : Entre 3 et 6 occurences de a
\b : les frontières d'un mot
\B : les non frontières d'un mot
^ : Début de la chaine de caractère
$ : Fin de la chaine de caractère'''

#Utiliser le module "re":
import re
text = "Une planète est un corps céleste orbitant autour du Soleil ou d'une autre étoile, possédant une masse suffisante pour que sa gravité la maintienne en équilibre hydrostatique, c'est-à-dire sous une forme presque sphérique, et ayant éliminé tout corps « rival » se déplaçant sur son orbite ou sur une orbit "

#re.search(texte_a_chercher,texte) Cherche tout la chaine de caractère, return avec .start() et .end() les indexs des recherches
#re.match(texte_a_chercher,texte) Cherche au début de la chaine de caractère
#return un  MatchObject si ok ou None si pas trouvé
#print(re.search("planète", text))
#print(re.search("tigre", text))

#re.compile(text, args): permet de compiler (créer) des expressions régulières
test = re.compile("^une", )
print(test.search(text))

# example code:
text2 = """Test pour
Tester"""

import re

print(re.match('Test', text2)) #OK
print(re.match('Tester', text2)) #None
print(re.match('^Tester', text2)) #None
print(re.search('Tester', text2)) #OK
print(re.search('^Tester', text2, re.MULTILINE)) #OK

#re.findall(regex, texte) retourne un match dans une liste[]
# ou re.search().group() (différence entre re.findall et re.search pas trop comprise par personne)
test2 = re.findall("une", text) #Cherche tous les "une" dans la chaine de caractère et convertie en liste
test3 = re.search("une", text) #Cherche le premier "une" dans la chaine de caractère
test4 = re.search("une", text).group() #Cherche le premier "une" dans la chaine de caractère etr retourne "une"
print(test2)
print(test3)
print(test4)

#re.split('regex', text)
test5 = re.split("une", text) #Split toute la chaine de caractère sur "une" et créé une liste
test6 = re.split("[A-Z]", text) #Split sur toutes les majuscules de toutes les lettres de l'alphabet "[A-Z]" étant un token
print(test5)
print(test6)
```

#**III- Les modules, Packages et Librairies**
Bibliothèque = Library(biblio en anglais) = Libraries (biblio au pluriel en anglais)

```python
# I - LES MODULES : fichier ".py" qui comporte une ou plusieurs fonctions.

'''Il y a deux façon d'importer des modules :'''

  #1) Importer tout le module via "import":

import operations #On importe de tout le module "operations"

result = operations.sum(2,5) #On appelle la méthode ".sum" du module "operations" et ses arguments "(2,5)"
print(result)

  #2.1) Importer une partie de module via "from" puis "import":

from operations import sum #On importe uniquement la méthode ".sum"

result = sum(2,5) #On appelle uniquement la méthode "sum" du module "operations"
print(result)

  #2.2) Importer plusieurs parties du module via "from" et "import":

from operations import sum, division #On importe les méthodes ".sum" ".division"

result = division(2,5) #On appelle uniquement une des méthodes "sum" ou "division" du module "operations"
print(result)

  # 3) Création de modules :

'''On créé un fichier. Exemple : "fonction.py"
On met notre fonction dans ce fichier. Exemple : def carre(nbr):
                                                     resultat = nbr**2
                                                     return resultat

Dans notre fichier de travail on importe notre fonctiontion :

from fonction import carre'''

```
