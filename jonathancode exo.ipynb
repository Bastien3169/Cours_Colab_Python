```python
def rep_num(question):
    print(question[0])
    for i in range(len(question[1])):
      print(i+1, "-", question[1][i])
    try:
      reponse=int(input("Ecrire entre 1 et " + str(len(question[1])) + " "))
      if 1 <= reponse <= len(question[1]):
        return reponse-1
      else:
        print("Il n'y a pas ce choix!!!!")
        return rep_num(question)
    except ValueError:
      print("Ce n'est pas un nombre!!!!")
      return rep_num(question)

def questionnaire(question):
  if question[1][rep_num(question)] == question[2]:
    print("Réponse juste!")
  else :
    print("Réponse fausse!")

question1 = ("Quelle est la capitale de la France: ", ("Paris", "Londres", "Toulouse", "Cherbourg"), "Paris")
question2 = ("Quelle est la capitale de l'Angleterre: ", ("Paris", "Londres", "Toulouse", "Cherbourg"), "Londres")
questionnaire(question1)
questionnaire(question2)
```

```python
def rep_num(question):
    print(question[0])
    for i in range(len(question[1])):
        print(i + 1, "-", question[1][i])
    try:
        reponse = int(input("Ecrire entre 1 et " + str(len(question[1])) + " "))
        return reponse - 1
    except ValueError:
        print("Ce n'est pas un nombre!!!!")
    except IndexError:
        print("Il n'y a pas ce choix!!!!")
        return rep_num(question)

def questionnaire(question):
    try:
        if question[1][rep_num(question)] == question[2]:
            print("Réponse juste!")
        else:
            print("Réponse fausse!")
    except IndexError:
        print("Indice invalide!")

question1 = ("Quelle est la capitale de la France: ", ("Paris", "Londres", "Toulouse", "Cherbourg"), "Paris")
question2 = ("Quelle est la capitale de l'Angleterre: ", ("Paris", "Londres", "Toulouse", "Cherbourg"), "Londres")
questionnaire(question1)
questionnaire(question2)
```
