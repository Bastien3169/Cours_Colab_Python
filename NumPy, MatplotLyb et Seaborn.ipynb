#**I - NUMPY : TABLEAU 1D ET 2D**

```python
# I - TABLEAU 1D

import numpy as np

#!!!!!! Il est imppssible de comparer une liste avec une valeur en Python.
#!!!!!! En revanche, il est possible d'effectuer une comparaison directe sur l'array.

# Liste python simple :
revenus = [1800, 1500, 2200, 3000, 2172]

# Création d'un "array" :
revenus_array = np.array(revenus)

'''
Si nous voulion faire une moyenne de la liste suivante, ou une comparaison de liste :
il n’y a pas de fonction existante par défaut permettant de le faire, il faudrait :
- la recoder soi-même !
- C’est encore plus compliqué si on souhaite avoir une information plus spécifique, comme la médiane.
- Faire une focntion par itération pour faire une comparaison de salaire (print(revenus>200) ne fonctionne pas)

Il faut donc créer un "array" comme suit :

revenus = [1800, 1500, 2200, 3000, 2172]
revenus_array = np.array(revenus)
'''

# Les arrays NumPy : similaire à une liste Python mais pas de comparateur (<;>; +; ==; -; etc)
  # Sans array (liste Python normale)
print("----- sans array -----")
print(sum(revenus) / len(revenus))
  # Avaec array
print("----- avec array -----")
print(np.mean(revenus))


# Quelques méthodes array 1D :
print("----- Quelques méthodes array -----")
print(np.zeros(4)) # 4x0
print(np.ones(4)) # 4x1
print(np.arange(2, 12, 2)) # de 2 à 12 avec un pas de 2.
print(np.linspace(2, 6, 4)) # 4 valeurs qui se suivent entre 2 et 6


# Selection dans l'array via les slices array, idem python :
print("----- Les slices array, idem python -----")
print(revenus[4]) # pour accéder au 5ème élement
print(revenus[-1]) # pour accéder au dernier élément
revenus[1] = 1900 # On peut aussi modifier les valeurs
print(revenus[1])

'''
nom_array[i:j:p]:

- i : l’indice du début
- j : l’indice de fin
- p : le pas
'''
print(revenus[:3]) # Les 3 premiers éléments
print(revenus[2:]) # Les éléments à partir de l’indice 2
print(revenus[::2]) # Un élément sur deux
print(revenus[::-1]) #Si le pas est négatif, le début et la fin du slice sont inversés. Utiliser pour inverser un tableau.


# Selection dans l'array via une condition :
'''
On pourrait naïvement exécuter revenus_array[ > 2000] en pensant que Python comprendra
qu’on souhaite appliquer la condition à l’array qui est cité juste avant. Le résultat sera une erreur.
Il faut bien considérer que l’array dans lequel on sélectionne, et la condition à appliquer,
sont deux choses différentes. Même si cela concerne le même array, il est nécessaire de le spécifier
également dans la condition !
'''
print("----- impossible de comparait sur une liste, il faut transformer en array-----")
print("revenus[revenus > 2000] ne fonctionne pas")

print("----- trasformation en array -----")
print(revenus_array[revenus_array > 2000])

print("----- trasformation en liste -----")
print(revenus_array[revenus_array > 2000].tolist())


# Remplacer un élèment dans l'array :
print("----- Remplacer un élèment dans l'array -----")
revenus_array[0]=21 # Je remplace le chiffre de l'indice [0] par 21
print("Remplacement de l'indice [0] par 21 :", revenus_array)
revenus_array[revenus_array > 2000] = 0 # Je remplace tous les éléments >2000 dans l'array par 0
print("remplacement de tous les éléments >2000 :", revenus_array)
revenus_array[revenus_array == 2172] = 3172 # Je remplace 2172 dans l'array par 3172
print("Remplacement 2172 dans l'array par 3172 :", revenus_array)



# II - TABLEAU 2D

'''
Voici à quoi ressemblerai un tableau avec des listes python... Pas à grand chose
hugo = [21, 1400, 0]
richard = [54, 2800, 2]
emilie = [27, 3700, 3]
tableau = [hugo, richard, emilie]
'''

# Quelques méthodes array 2D :

display(np.zeros((4, 4))) # 4x4 rempli de 0
display(np.ones((3, 5))) # 3x5 rempli de 1
display(np.random.random((6, 3))) # 6x3 rempli de valeurs aléatoires comprises entre 0 et 1
display(np.random.randint(1, 10, size=(3, 3))) # 3x3 rempli de valeurs aléatoires entières, comprises entre 1 et 10


# Exemple de fonctionnement tableau array :

  # Création de listes:
hugo = [1800, 21, 0]
richard = [1500, 54, 2]
emilie = [2200, 28, 3]
pierre = [3000, 37, 1]
paul = [2172, 37, 2]
deborah = [5000, 32, 0]
yohann = [1400, 23, 0]
anne = [1200, 25, 1]
thibault = [1100, 19, 0]
emmanuel = [1300, 31, 2]

  # Création de listes de liste:
tableau = [hugo, richard, emilie, pierre, paul, deborah,
           yohann, anne, thibault, emmanuel]

  # Tableau Python:
display(tableau)

  # Création de l'array:
array = np.array(tableau)

  # Faire apparaitre la 1ère ligne entière:
print("----- 1ère ligne entière -----")
display(array[0, :])
print("----- salaires > 2000 en python -----")
display(array[ :,0] > 2000)
print("----- salaires > 2000 en array -----")
display(array[array[ :,0] > 2000])

  # Ajouter une nouvelle personne au tableau:
print("----- Ajout valeur : np.vstack(array, valeur) -----")
bast = [2750, 40, 1]
display(np.vstack((array, bast)))

  # .reshape = Reshape un array :
```

```python
# Création de la limitation du schéma ou seront les 4 grpahs
plt.figure(figsize=(12, 12))

# 1) Proportion de prêt par type de prêt : cammenbert
plt.subplot(2, 2, 1) # crée grille de 2 lignes et 2 colonnes, et place le graphique dans la 1ère position.
plt.pie(x = df["type"].value_counts(), labels = ["Immobilier", "Automobile"], autopct = '%1.1f%%')
plt.title("Proportion de prêt par type de prêt")


# 2) Bénéfice mensuel réalisé en fonction du revenu du client - prêts immobiliers : nuage de points
plt.subplot(2, 2, 2)
pret_immo = df.loc[df["type"] == "immobilier", : ] # Création series colonne immobilier
plt.scatter(pret_immo["revenu"], pret_immo["benefices"])
plt.title("Bénéfice mensuel réalisé en fonction du revenu du client")


# 3) Distribution des bénéfices réalisés : histogramme
plt.subplot(2, 2, 3)
plt.hist(df["benefices"], rwidth = 0.8)
plt.title("Distribution des bénéfices réalisés")


# 4) Bénéfice mensuel total réalisé par agence : barre
plt.subplot(2, 2, 4)
benef_ville = df.groupby("ville")["benefices"].sum()
plt.bar(height = sorted(benef_ville), x = df["ville"].unique())
plt.title("Bénéfice mensuel total réalisé par agence")

# On montre ce shéma des 4 graphs
plt.show()
```

#**V.d - LES LIGNES OU COURBES (.lineplot() : variable num = f(tps))**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/main/data/prets_final.csv')

# Données
'''
x = variable_non_num
y = variable_num
hue = rajoute une sous-catégorie pour comparer 2 valeurs sur une barre par couleur
estimator = si pas indiqué, par defaut 'np.mean' = calcul pour représentation des barres
ci = si pas indiqué, par defaut '95' = représente la barre d'erreur
orient = oriantation graphique "v" = verticale, "h" = horizontale
palette = choix des couleurs des barres pour "hue"
saturation = saturation des couleurs des barres
order = ['Paris', 'Lyon', 'Toulouse', 'Marseille', 'Bordeaux'] = ordre personnalisé pour l'affiche des 'x'
hue_order = ['automobile', 'immobilier'] = ordre personnalisé pour l'affiche des sous-catégories
style = permet d'avoir un style de ligne différent pour la colonne spécifiée. Fonctionne avec "markers", "dashes", etc
markers = met des marqueurs différents pour les points de chaque ligne grace à l'argument "style"
dashes = met des style de ligne différents pour chaque ligne grace à l'argument "style"
'''

sns.lineplot(data=df, x='ville', y='revenu', hue="type", style="type", markers=True, dashes= True,
             estimator='sum', ci=None, orient = "x", palette="Greens", legend="auto", linewidth=4)

#estimator=sum
plt.show()
```

#**V.e - ASPECT GRAPHIQUE ET NOTION DE GLOBAL ET LOCAL**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/main/data/prets_final.csv')

# Données globales
plt.figure(figsize=(12, 12)) # Global : pour tous les graphiques sauf si on précise dans l'un d'eux en argument, ce qui annulera pour celui-ci la taille
sns.set_theme(style='whitegrid', palette='pastel') # Global : pour tous les graphiques
sns.set_palette('Set2') # Global : pour tous les graphiques, écrit après donc écrase "palette='pastel'" de '.set_theme'

# Graphique 1 : global
sns.lineplot(data=df, x='ville', y='revenu', hue="type", style="type", ci=None) # Ici on se sert de 'plt.figure', '.set_theme' et '.set_palette' comme global

# Graphique 2 : local. Ici, l'argument 'palette = "Set1"' écrase le paramètre global 'sns.set_palette('Set2')' pour ce graphique uniquement.
plt.figure(figsize=(9, 9)) # Ici plt.figure écrase le 1er
sns.lineplot(data=df, x='ville', y='revenu', hue="type", style="type", ci=None, palette = "Set1") # Ici on se sert de 'plt.figure', '.set_theme' et '.set_palette' comme global

# Montrer graphique
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/main/data/prets_final.csv')

# Données globales
sns.set_theme(style='whitegrid', palette='Set2', rc={"lines.linewidth": 3, "lines.color": "pink"})

# Graphique 1 : global
sns.lineplot(data=df, x='ville', y='revenu', hue="type", style="type", ci=None) # Ici on se sert de 'plt.figure', '.set_theme' et '.set_palette' comme global

# Montrer graphique
plt.show()
```

#**V.f - sns.set_theme(rc={"figure.facecolor": "white"})**

```python
'''
1) FIGURE :
- figure.figsize: Taille par défaut de la figure (par exemple, [8, 6]).
- figure.facecolor: Couleur de fond de la figure.
- figure.edgecolor: Couleur du bord de la figure.

2) LIGNES :
- lines.linewidth: Largeur des lignes.
- lines.color: Couleur des lignes.
- lines.linestyle: Style des lignes (par exemple, '-', '--', ':').

3) AXES :
- axes.titlesize: Taille du titre des axes.
- axes.labelsize: Taille des étiquettes des axes.
- axes.grid: Affichage de la grille (booléen).
- axes.facecolor: Couleur de fond de la zone des axes.

4) TICKS (graduations) :
- xtick.color: Couleur des ticks sur l'axe des x.
- xtick.direction: Direction des ticks sur l'axe des x ('in', 'out').
- ytick.color: Couleur des ticks sur l'axe des y.
- ytick.direction: Direction des ticks sur l'axe des y ('in', 'out').

5) LEGENDE :
- legend.fontsize: Taille de la police dans la légende.
- legend.loc: Position de la légende ('best', 'upper right', 'lower left', etc.).
- legend.frameon: Affichage ou non du cadre autour de la légende.

6) POLICE :
- font.size: Taille globale de la police.
- font.family: Famille de la police (par exemple, 'serif', 'sans-serif', 'monospace').
- font.serif, font.sans-serif, font.monospace: Listes de polices spécifiques pour ces catégories.

7) COULEURS :
- text.color: Couleur du texte.
- axes.labelcolor: Couleur des étiquettes des axes.
- xtick.color, ytick.color: Couleur des ticks.
'''

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/OpenClassrooms-Student-Center/fr-4452741-decouvrez-les-librairies-python-pour-la-data-science/main/data/prets_final.csv')

# Données globales
sns.set_theme(rc={"figure.figsize": [10, 8], "figure.facecolor": "white", "figure.edgecolor": "blue",
                  "lines.linewidth": 2, "lines.color": "green", "lines.linestyle": "--",
                  "axes.titlesize": 16, "axes.labelsize": 14, "axes.grid": True,
                  "xtick.color": "black", "xtick.direction": "out","ytick.color": "green", "ytick.direction": "in",
                  "legend.fontsize": 12, "legend.loc": "best", "legend.frameon": False,
                  "font.size": 12, "font.family": "serif",
                  "text.color": "black", "axes.labelcolor": "blue", "xtick.color":"pink"
                  })

# Graphique 1 : global
sns.lineplot(data=df, x='ville', y='revenu', hue="type", style="type", ci=None) # Ici on se sert de 'plt.figure', '.set_theme' et '.set_palette' comme global

# Montrer graphique
plt.show()
```
