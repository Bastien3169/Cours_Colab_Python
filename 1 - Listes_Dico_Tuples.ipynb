# **II- LES LISTES**

```python
# 1) ACCEDER/MODIFIER UNE LISTE :
liste = ["actions", "obligations", "immobilier"]

#Accéder à la valeur avec l'indice
print(liste[0])

#Accéder à une plage de valeur avec "slice"
print(liste[0:2]) #On peut mettre un pas liste[0:2:2]

#Modifier (ou remplacer) les éléments d'une liste:
liste[-2] = 15 #Remplace "immobilier" par "15"
print(liste)

#Créer un liste avec "range()"
liste2=list(range(1,5))
print(liste2)

#Boucler (ou itérer sur une liste)
print("--------")
for i in range(len(liste)):
    print(liste[i])
```

```python
# 2) AJOUTER DES ELEMENTS A UNE LISTE :
liste = ["actions", "obligations", "immobilier"]

#.append: Ajoute l'élément à la fin d'une liste:
liste.append("snow")
print(liste)

#.insert: Ajoute l'élément à l'endroit donné par l'indice:
liste.insert(2, "foot")
print(liste)
#OU
#slice : Ajoute l'élément à l'endroit donné par l'indice:
liste[1:1] = ["skate"]
print(liste)

# + : Ajoute une liste à une autre liste:
liste_bis = ["crypto", "produit struct"]
liste_tot = liste_bis + liste
print(liste_tot)

#.extend: Rajoute une liste (plusieurs éléments) à une liste à la fin
liste.extend([12, 15])
print(liste)
```

```python
# 3) FONCTION "zip()" : CREER UNE LISTE A PARTIR DE DEUX LISTES AVEC "list" :

'''Permet de créer à partir de 2 listes :
  - Soit un dictionnaire avec le constructeur "dict".
  - Soit une liste de tuple avec le constructeur "list".
 avec : l'élément 1 : les clés et l'éléments 2 : les valeurs'''

age = [12, 23, 67, 42]
prenom = ["Anais", "Aubin", "Bastien", "Maman"]

print(zip(age, prenom)) # Crée un objet zip donc pas printable
print(list(zip(age, prenom))) # On transforme avec le constructeur "list" pour printer une liste de tuples
print(dict(zip(age, prenom))) # On transforme avec le constructeur "dict" pour printer un dictionnaire
```

```python
# 4) SUPPRIMER LES ELEMENTS A UNE LISTE :
liste = ["actions", "obligations", "immobilier", "liquidité"]

#.remove : supprime l'élément pointé par la valeur (et non l'indice) mais retourne "none"
liste.remove("liquidité")
print(liste)

#.pop : supprime l'élément pointé par l'indice ou le dernier si pas d'indice (et non la valeur) et retourne la valeur supprimée.
liste.pop(-1)
print(liste)

#del : Supprime l'élément pointé par l'indice:
del liste[-1]
print(liste)

#.clear : supprime une liste entière:
liste.clear()
print(liste)
```

```python
# 5) AUTRES METHODES POUR UNE LISTE
liste = [1, 1, 2, "trois", 4, "cinq", 6]
liste2 = [3, 6, 2, 1]

#len(): Fonctrion pour connaitre la longueur d'un liste avezc "len()" et surtout permet de mettre des numéros à chaque élement d'une liste et de jouer avec:
longueur=len(liste)
print("Longueur de la liste:",longueur)
  #Exemple d'utilisation de "len()":
'''for i in range(len(liste)):
  print(i)'''
#len() : Connaitre la longueur d'un indice dans une liste avec "len":
longueur=len(liste[3])
print("Longueur de l'indice 3 dans la liste (trois) :", longueur)

#.slice : Récupérer des tranche (ou slice) dans une liste si pas de chiffre, prend depuis le début et/ou la fin:
tranche = liste[2:4]
print("Sans pas :", tranche)
tranche = liste[0:6:2]
print("Pas de 2 (un sur deux) :", tranche)
tranche = liste[::-1]
print("Pas -1 (inverse la liste) :", tranche)

#.sort : Trie une liste avec "sort": Trie une liste par ordre alphabétique ou numérique (attention les majuscules sont prios au minuscules):
liste2.sort()
print("Trie la liste :", liste2)

#.reverse : Inverser l'ordre des éléments d'une liste avec "reverse":
ordre=liste2.reverse()
print("Inverse la liste :", ordre)

#.index : Renvoyer la position de la valeur en indice avec "index":
print("'trois' est à la position :", liste.index("trois"))

#.count : Renvoie le nbre d'occurence de l'élément sépcifier dans un str, un int ou une liste attention c'est les caractères et non les mots:
print("L'occurence du chiffre '1' ds cette liste est de :", liste.count(1))
  #Autre exemple :
liste3 = "Bonjour le jour"
print("L'occurence de 'jour' ds ce str est de :", liste3.count("jour")) #2 occurence car pas d'espace et donc 2 fois "jour"
print("L'occurence de 'jour' ds ce str est de :", liste3.count(" jour ")) #1 occurence car des espaces et donc un seul " jour "

#Savoir si un élément est présent dans la liste
print("'5' est-il dans la liste ?", 5 in liste)

#.split : spliter, pour que chaque élément devienne une liste. Attention uniquement q'un seul argument:
  #Attention on ne peut pas "spliter" sur une liste. Donc:
liste_bis= "1, 1, 2, trois, 4, cinq, 6"
separe=liste_bis.split(",")
print(separe)

#.join : join des listes et fonctionne bien avec ".split" :
liste_bis2= "1,2"
print("Joindre via .split :" , ("-").join(liste_bis2.split(",")))
  #Ou directement sur une liste de chaine de caractère :
liste_bis3 = ["1","2"]
print("Joindre une liste avec chaine caratère :" , ("-").join(liste_bis3))

#.copy : Copie une liste entière dans une autre variable
liste_copy = liste.copy()
print(".copy :", liste_copy)

# Parcourir une liste bidimensioelle (liste dans une liste) avec une boucle imbriquée :
listes2 = [["BTC", "ETH"], ["Euro", "Dollar"], ["Or", "Argent"]]

for i in listes2 :
  for ii in i :
    print(ii)
```

```python
# 6) FONCTIONS AVEC LES LISTES DE LISTES:

# Liste de listes
liste= [["Bast", "Anais", "Aubin"], ["JP", "Peter", "Chris"], ["Dupont", "Ntamak", "Beile"]]

print("--- Liste originale ---")
print(liste[0])

print("--- Modifier un élément dans une sous-liste ---")
liste[0][1] = "Nanou"
print(liste[0])

print("--- Ajouter un nouvel élément à la fin d'une sous-liste ---")
liste[1].append("Devun")
print(liste[1])

print("--- Ajouter une nouvelle sous-liste ---")
liste.append(["moi", "remoi", "moimoi"])
print(liste)

print("--- Modifier une sous-liste entière ---")
liste[2] = ["snow", "skate", "wake"]
print(liste[2])

print("--- Supprimer un élément dans une sous-liste ---")
del liste[1][2]
print(liste[1])

liste[1].pop(2) # Avec indice
print(liste[1])

liste[1].remove("Peter") # Avec valeur
print(liste[1])

print("--- Supprimer une sous-liste ---")
del liste[3]
print(liste)
```

#**III- LES DICTIONNAIRES**

```python
# 1) CREER/ACCEDER/MODIFIER UN DICTIONNAIRE :

#Voici un dictionnaire avec ses "clés":"valeurs":
sport = {"accrobatique":"snow", "endurance":"trail", "collectif":"foot"}
print(sport)

# a) Créer un dictionnaire vide avec "{}" comme une liste vide "[]"
  #et on y joute les éléments "clé":"valeur" normalement
dico_sport={}
dico_sport["accrobatique"]="snow"
dico_sport["endurence"]="trail"
dico_sport["collectif"]={"mains":"hand"}
print(dico_sport)
  #ou avec "dict()"
dico_sport=dict()
dico_sport["accrobatique"]="snow"
dico_sport["endurence"]="trail"
dico_sport["collectif"]= {"mains":"hand"}
print(dico_sport)

# b) Acceder à une valeur via une clé.
print(dico_sport["accrobatique"])
print(dico_sport["collectif"]["mains"]) # Accéder à la clé d'une clé d'une valeure
  #Ou avec ".get" moins friendly mais retourne "none" si pas de valeur
print(dico_sport.get("accrobatique"))

# d) Modifier une valeur associé à une clé:
dico_sport["collectif"]="basket"
print(dico_sport)
```

```python
# 2) AUTRES METHODES D'ACCES D'UN DICTIONNAIRE :

sport = {"accrobatique":"snow", "endurance":"trail", "collectif":"foot"}

#.items() : Retourne une vue des clés:valeurs du dico :
print("-----.items() : retourne une vue des clés:valeurs-----")
print(sport.items()) #dict_items
print(list(sport.items())) # donne une liste de tuple

#.key() : Retourne une vue sur les clés du dico :
print("-----.key(): Retourne une vue sur les clés-----")
print(sport.keys()) #dict_keys
print(list(sport.keys()))# donne une liste

#.values() : Retourne une vue sur les valeurs du dico :
print("-----.values() : Retourne une vue sur les valeurs-----")
print(sport.values()) #dict_values
print(list(sport.values()))# donne une liste

#.copy() : Copiez l'intégralité du dictionnaire dans un nouveau dictionnaire
print("-----.copy() : Copiez l'intégralité du dictionnaire dans un nouveau dictionnaire-----")
sport2 = sport.copy()
print(sport2)
```

```python
# 3) AJOUTER DES ELEMENTS A UN DICTIONNAIRE :

sport = {"accrobatique":"snow", "endurance":"trail", "collectif":"foot"}
sport2 = {"Bastien":"père", "Anais":"maman", "Aubin":"fils"}

#.update : Ajout d'une clé:valeur. Attention, créé la nouvelle clé:valeur à la fin. Moins performent, moins lisible mais plus pratique si ajout plusieur clé/valeurs
print("-----.update()-----")
sport.update({"solo":"tennis"}) #Equivalent ".append" des listes.
sport.update({"solo":"tennis","seul":"voile"}) # On peut ajouter plusieurs clés/valeurs en un coup.
sport.update({"solo":{"raquette":"tennis"}}) # On peut imbriquer des dicos.

#.update() : copie un dictionnaire dans un autre dictionnaire.
sport.update(sport2)
print("Ceci est un .update() de dico :", sport)

print("-------------- sport['e-sport']={'plateforme':'mario','rpg':'zelda'} ------------")
#Créer d'un coup, plusieur clés/valeurs dans un dico.
sport["e-sport"]={"plateforme":"mario","rpg":"zelda"}
print(sport)

#Créer les clés/valeurs une à une :
sport["e-sport"]["plateforme"]="sonic"
print('["e-sport"]["plateforme"]="sonic" : ', sport)
sport["e-sport"]["rpg"]="secret of mana"
sport["e-sport"]="fifa" #On ne peut en rajouter que un par un MAIS...
sport["bizarre"]={"longe-côte":"marche"} #... on peut imbriquer des dicos.
print(sport) #Si on ajout un entier pas besoir de ""

#zip() : Fonction créée un dictionnaire à partir de 2 listes :
'''Permet de créer à partir de 2 listes :
  - Soit un dictionnaire avec le constructeur "dict".
  - Soit une liste de tuple avec le constructeur "list".
 avec : l'élément 1 : les clés et l'éléments 2 : les valeurs'''

age = [12, 23, 67, 42]
prenom = ["Anais", "Aubin", "Bastien", "Maman"]

print(zip(age, prenom)) # Crée un objet zip donc pas printable
print(list(zip(age, prenom))) # On transforme avec le constructeur "list" pour printer une liste de tuples
print(dict(zip(age, prenom))) # On transforme avec le constructeur "dict" pour printer un dictionnaire
```

```python
# 3.1) EXERCICES AJOUT DICTIONNAIRE

text = "um sed propinqua omnia omnia se caedium campestrem propinqua parte parte."
liste_text = text.split() # .split() rend text en "list". Par défaut si pas d'argument ds .split(), split aux espaces

liste_occurence = {} # Création dictionnaire vide
for i in liste_text:
    if i not in liste_occurence :
        liste_occurence.update({i:liste_text.count(i)}) # 1ere façon : moins performent, moins lisible mais plus pratique si ajout plusieurs clé/valeurs.
        liste_occurence[i] = liste_text.count(i) # 2eme façon : plus intuitive et concise, modifie le dictionnaire existant sans nécessiter de création temporaire.
    else :
        pass
print(liste_occurence)
```

```python
# 4) SUPPRIMER DES ELEMENTS A UN DICTIONNAIRE :

sport = {"accrobatique":"snow", "endurance":"trail", "collectif":"foot", "réflexion":"échec"}

#.pop() : Supprime une valeur associé à une clé retourne l'élément supprimé
print(sport.pop("collectif")) # Print uniquement la valeur de la clé supprimer
print(sport)

#del : Supprime une valeur associé à une clé définitivement, moins friendly mais retourne "none" si pas de valeur
del sport["endurance"]
print(sport)

#.popitem() : Supprime la dernière valeur ajoutée
print(sport.popitem())
print(sport)

#.clear() : Supprimer un dictionnaire entier
sport.clear()
print(sport)
```
