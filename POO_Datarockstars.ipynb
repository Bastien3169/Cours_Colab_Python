```python
# Projet fiche de profil

class DataProfile:
    def __init__(self, name, skills):
        self.name = name
        self.skills = skills

    def display_skills(self):
        print("----------")
        print(f"Compétences de {self.name}:")
        for skill, level in self.skills.items():
            print(f"{skill}: {level}/10")

class DataEngineer(DataProfile):
    pass

class DataScientist(DataProfile):
    pass

class DataAnalyst(DataProfile):
    pass

# Création des profils avec des compétences fictives
engineer = DataEngineer("Data Engineer", {"Hadoop": 7, "Spark": 8, "Kafka": 6, "ETL": 9, "SQL": 8})
scientist = DataScientist("Data Scientist", {"Machine Learning": 9, "Statistiques": 8, "Deep Learning": 7, "Python": 9})
analyst = DataAnalyst("Data Analyst", {"Analyse de Données": 8, "SQL": 9, "Tableau": 9, "Excel": 9})

# Affichage des compétences
engineer.display_skills()
scientist.display_skills()
analyst.display_skills()
```

```python
# Projet Ajout/suppresion ou recherche sur carnet de contact


# Liste de dictionnaires carnet de contacts
carnet_de_contacts = [
    {"Nom": "Dupont", "Prénom": "Antoine", "Mail": "antoine.dupont@st.com"},
    {"Nom": "Zinedine", "Prénom": "Zidane", "Mail": "Zidane.einedine@fr.com"},
    {"Nom": "Riner", "Prénom": "Teddy", "Mail": "teddy.riner@judo.com"}
    ]



# Suppression d'un contact par son nom ou prénom
def supression():
  reponse = int(input("Voulez-vous supprimer par : \n 1/ Par nom \n 2/ Par prénom \n Taper 1 ou 2 puis 'Entrée'. "))
  print("----------")

  if reponse == 1:
    nom_a_supprimer = input("Ecrire le nom à supprimer = ").capitalize()
    print("----------")
    for i in carnet_de_contacts:
        if i["Nom"] == nom_a_supprimer:
            carnet_de_contacts.remove(i)
            print(f"Le contact {nom_a_supprimer} a été supprimé.")
            break
    else:
        print(f"Aucun contact trouvé avec le nom {nom_a_supprimer}.")
        print("----------")

  elif reponse == 2:
    prenom_a_supprimer = input("Ecrire le prénom à supprimer = ").capitalize()
    print("----------")
    for i in carnet_de_contacts:
        if i["Prénom"] == prenom_a_supprimer:
            carnet_de_contacts.remove(i)
            print(f"Le contact {prenom_a_supprimer} a été supprimé.")
            break
    else:
        print(f"Aucun contact trouvé avec le prénom {prenom_a_supprimer}.")
        print("----------")

  print("Carnet de contacts mis à jour :") # Affichage du carnet de contacts mis à jour
  for i in carnet_de_contacts:
      print(i)



# Ajouter un contact
def ajout():
  nouveau_contact = {
      "Nom": input("Ecrire le nom du nouveau contact = ").capitalize(),
      "Prénom": input("Ecrire le prénom du nouveau contact = ").capitalize(),
      "Mail": input("Ecrire le mail du nouveau contact = ").capitalize(),
      }
  print("----------")
  carnet_de_contacts.append(nouveau_contact)
  print("Carnet de contacts mis à jour :")
  for i in carnet_de_contacts:
      print(i)



# Chercher un contact par son nom
def cherche():
  reponse = int(input("Voulez-vous chercher par : \n 1/ Par nom \n 2/ Par prénom \n Taper 1 ou 2 puis 'Entrée'. "))
  print("----------")

  if reponse == 1:
    nom_a_chercher = input("Ecrire le nom à chercher = ").capitalize()
    print("----------")
    for i in carnet_de_contacts:
        if i["Nom"] == nom_a_chercher:
            print(i)

  if reponse == 2:
    prenom_a_chercher = input("Ecrire le nom à chercher = ").capitalize()
    print("----------")
    for i in carnet_de_contacts:
      if i["Prénom"] == prenom_a_chercher:
        print(i)


# Question de départ
reponse = int(input("Voulez-vous : \n 1/ Chercher un contacte ? \n 2/ Ajouter un contacte ? \n 3/ Supprimer un contacte ? \n Taper 1, 2 ou 3 puis 'Entrée'. "))
print("----------")

if reponse == 1:
    print("Chercher un contacte")
    cherche()
elif reponse == 2:
    print("Ajouter un contacte")
    ajout()
elif reponse == 3:
    print("Supprimer un contacte")
    supression()
```

```python
# GERER LES FICHIERS

# I- OUVERTURE D'UN FICHIER : 2 façons d'ouvrir un fichier

  # 1) with open :
'''
Méthode automatique et recommandée pour lecture de fichier :
 - Le fichier est automatiquement fermé dès que le bloc with se termine, pas besoin d'appeler explicitement fichier.close().
 - Plus concis, plus clair mais surtt plus fiable pour éviter des problèmes comme un fichier qui reste ouvert par accident.
 - Moins de code manuel à écrire
'''
  # Exemple avec tous les arguments:
with open(
    "mon_fichier.txt", # Chemin
    mode="r", # Mode lecture :
    encoding="utf-8", # Norme de décodage ("utf-8", "latin-1", "ascii", ...)
    buffering=1024, # On s'en fou : en octet
    errors="replace", # On s'en fou : # On s'en fou : Spécifie la façon de gérer les erreurs d'encodage/décodage.
    newline="\n" # On s'en fou : Contrôle la gestion des caractères de nouvelle ligne (\n, \r, \r\n).
) as fichier:
    contenu = fichier.read()
    print(contenu)

  # Exemple simple:
with open("mon_fichier.txt", "r", encoding="utf-8") as fichier:
    contenu = fichier.read()
    print(contenu)


  # 2) Mode normal :
'''
Méthode manuelle (normale) et non recommandée pour lecture de fichier :
 - Il faut manuellement fermer le fichier avec fichier.close()
 - Facile d'oublier d'appeler fichier.close(), surtout dans des codes complexes, ce qui gaspille des ressources système.
'''
fichier = open("mon_fichier.txt", "r", encoding="utf-8")
contenu = fichier.read()
print(contenu)
fichier.close()


# 3 ) Les différents modes :

'''
- "r" : Lecture seule (par défaut).
- "w" : Écriture seule (écrase le fichier s'il existe ou le crée sinon).
- "x" : Création et écriture. Lève une exception si le fichier existe déjà.
- "a" : Écriture seule. Ajoute les données à la fin du fichier sans écraser son contenu.
- "b" : Mode binaire (peut être combiné avec d'autres modes, ex. "rb" pour lecture binaire).
- "t" : Mode texte (par défaut, peut être combiné comme "rt").
- "+" : Mode lecture/écriture.
'''

# II- PRINCIPALES METHODE POUR LA MANIPULATION DES FICHERS

 # 1) Lecture de fichier
print("---- .read() ----") # Lit tout le contenu du fichier sous forme d'une seule chaîne de caractères.
with open("mon_fichier.txt", "r", encoding="utf-8") as fichier:
  contenu = fichier.read()
  print(contenu)


print("---- .readline() ----") # Lit une seule ligne à la fois à partir du fichier (économie mémoire tanpon pour gros fichier).
with open("mon_fichier.txt", "r", encoding="utf-8") as fichier:
  ligne = fichier.readline()
  print(ligne)


print("---- .readlines() ----") # Lit toutes les lignes et les met dans une liste.
with open("fichier.txt", "r") as source:
    lignes = source.readlines()
    print(lignes)


  # 2) Ecriture ou ajouter du contenu dans un fichier
print("---- .write('Hello World') ----") # Écrit une chaîne de caractères dans le fichier.
with open("fichier.txt", "w") as cible:
    cible.write("Bonjour tout le monde !")


print("---- .writelines([liste]) ----") # Écrit une liste de chaînes de caractères dans le fichier. Chaque chaîne correspond à une ligne.
with open("fichier.txt", "w") as cible:
    lignes = ["Ligne 1\n", "Ligne 2\n", "Ligne 3\n"]
    cible.writelines(lignes)


  # 3) Autre méthode
print("---- .seek(offset, whence) ----") # Change la position actuelle du curseur de lecture/écriture dans le fichier. "offset" : Nombre de bytes pour déplacer le curseur. "whence" : Point de référence (0 = début, 1 = position actuelle, 2 = fin)
with open("fichier.txt", "r") as source:
    source.seek(5)  # Déplace le curseur à partir du 5e caractère
    contenu = source.read()
    print(contenu)


print("---- .tell() ----") # Renvoie la position actuelle du curseur dans le fichier.
with open("fichier.txt", "r") as source:
    print(source.tell())  # Affiche la position actuelle


print("---- .close() ----") # Ferme explicitement le fichier (automatique avec with open).
source = open("fichier.txt", "r")
source.close()
```

```python
class Voiture:
    def __init__(self, marque, modele, couleur, kilometrage=0):
        self.marque = marque
        self.modele = modele
        self.couleur = couleur
        self.kilometrage = kilometrage

    def __str__(self):
        return (f"Voiture: {self.marque} {self.modele}, "
                f"Couleur: {self.couleur}, Kilométrage: {self.kilometrage} km")

    # Méthode pour changer la couleur
    def changer_couleur(self, nouvelle_couleur):
        self.couleur = nouvelle_couleur
        print(f"La couleur a été changée en {self.couleur}.")

    # Méthode pour mettre à jour le kilométrage
    def ajouter_kilometrage(self, kilometres):
        if kilometres < 0:
            print("Impossible d'ajouter un kilométrage négatif.")
        else:
            self.kilometrage += kilometres
            print(f"{kilometres} km ont été ajoutés. Nouveau kilométrage: {self.kilometrage} km.")

# Exemple d'utilisation
voiture1 = Voiture("Peugeot", "208", "Rouge", 10000)

print(voiture1)  # Affiche les informations initiales

# Modifier la couleur
voiture1.changer_couleur("Bleu")

# Ajouter du kilométrage
voiture1.ajouter_kilometrage(500)

# Ajouter un kilométrage négatif (test de validation)
voiture1.ajouter_kilometrage(-200)

print(voiture1)  # Affiche les informations mises à jour
```

```python
class Livre:
    def __init__(self):
        self.biblio = []

    def ajout_livre(self):
        print("Ajout d'un nouveau livre :")
        livre = {}
        livre["titre"]=input("Entrez le titre du livre : ")
        livre["Auteur"] = input("Entrez l'auteur du livre : ")
        livre["Année"] = input("Entrez l'année de publication : ")
        self.biblio.append(livre)
        print(self.biblio)


    def sup_livre(self):
        for i, y in enumerate(self.biblio, 1):
            print(f"{i} : {y}")

        titre_a_supprimer = input("Quel titre de livre voulez-vous supprimer ? ").lower()
        for i, y in enumerate(self.biblio):
            if y["titre"].lower() == titre_a_supprimer:
                del self.biblio[i]
                print(f"Le livre '{titre_a_supprimer}' a été supprimé, voici la nouvelle liste : ")
                print(self.biblio)
                break  # Sortir de la boucle si le livre est trouvé
        else:
            print("Le livre n'a pas été trouvé.")


    def affiche(self):
        for i, y in enumerate(self.biblio, 1):
            print(f"{i} : {y}")


biblio = Livre()

while True:
    print("\nMenu Bibliothèque :")
    print("1. Ajouter un livre")
    print("2. Supprimer un livre")
    print("3. Lister tous les livres")
    print("4. Quitter")

    choix = input("Choisissez une option (1-4) : ")

    if choix == "1":
        biblio.ajout_livre()
    elif choix == "2":
        biblio.sup_livre()
    elif choix == "3":
        biblio.affiche()
    elif choix == "4":
        print("Au revoir !")
        break
```
