#**I- Les fonctions les plus vues**

```python
liste = [1, 1, 2, "trois", 4, "cinq", 6]

# print():
print("______ print() ______")
"Besoin d'un dessin ? "
print("Ceci est un print d'une liste :", liste)

# input():
print("______ input()______")
"Besoin d'un dessin (bis) ? "
#input("ceci est un input ok ? ")

# int(), str(), float() et bool():
print("______ int(), str(), float() et bool() ______")
"Convertie l'argument en la fonction choisie"
print((int("3")), "3")

# list(), dict() et tuple():
print("______ list(), dict() et tuple() ______")
"Convertie l'argument en la fonction choisie"
test = 1, 2, 3
print("Il s'agit d'une variable de int :", test)
print("La variable de int est transformé en liste :", list(test))

# len():
print("______ len() ______")
" Renvoie le nombre d'éléments d'un 'str', d'une 'list', d'un 'dict', d'un 'tuple', etc..."
print("Longueur de la liste :",len(liste))
print("Longueur de l'indice 3 dans la liste (trois) :", len(liste[3]))

# slice():
print("______ slice() ______")
''' Indexe ou découpe une séquence comme une liste, un str, un dico, un tuple, etc... avec : len(start:stop:step)'''
print("Sans pas :", liste[2:4])
print("Pas de 2 (un sur deux) :", liste[0:6:2])
print("Pas -1 (inverse la liste) :", liste[::-1])

# range():
print("______ range() ______")
'''Génère une séquence d'entiers dans un intervalle donné avec : range(start, stop, step)'''
print(list(range(3)))
print(tuple(range(1,6,2)))

# min() et max():
print("______ min() et max() ______")
''' Obtient le plus petit et le plus grand élément d'une mêmê séquence (liste, tuple ou str (avec code ASCII ))'''
liste1 = 1, 2, 4, 6
liste2 = "trois", "quatre", "deux"
liste3 = "trois"
print("Print le 'int' le plus grand :", max(liste1))
print("Print le 'str' le + petit soit le 1er :", min(liste2))
print("Print la lettre du str la plus petite selon code ASCII :", min(liste3))

# count() :
print("______ count() ______")
print("Compte le nbr d'occurence de '1' : ", liste.count(1))

# round() :
print("______ round() ______")
'''Arrondit un nbr à un nbr spécifié de décimales :'''
print("J'arondie 3.145432 à 3 chiffres après la virgule :", round(3.145432, 3))

# index() : renvoie l'indice de la première occurrence de la valeur
print("______ index() ______")
print([7, 8, 9, 8].index(8)) # ici, retourne 1, car 8 apparaît pour la première fois à l'indice 1

# Starwith() : commence par... vérifie si une chaîne commence par un certain préfixe. Elle renvoie True si c'est le cas, sinon False.
print("______ Starwith() ______")
texte = "Bonjour tout le monde"
print(texte.startswith("Bonjour"))  # True
print(texte.startswith("bon"))   # False

# zip():
print("______ zip() ______")
'''Permet de créer à partir de 2 listes :
  - Soit un dictionnaire avec le constructeur "dict".
  - Soit une liste de tuple avec le constructeur "list".
 avec : l'élément 1 : les clés et l'éléments 2 : les valeurs'''

prenom = ["Anais", "Aubin", "Bastien", "Maman", "Papa", "Billy", "Caline"]

print(zip(liste, prenom)) # Crée un objet zip donc pas printable
print(list(zip(liste, prenom))) # On transforme avec le constructeur "list" pour printer une liste de tuples
print(dict(zip(liste, prenom))) # On transforme avec le constructeur "dict" pour printer un dictionnaire


# sorted() et .sort()
  # 1) sorted(): pour tout type d'objet (liste, dico, etc). Fonction ne pas confondre avec la méthode .sort()
print("______ sorted() ______")
'''
- retourne une nouvelle liste triée et laisse l'objet original inchangé.
- si vous avez besoin de conserver la liste d'origine intacte et que vous souhaitez obtenir une nouvelle liste triée.
- si on prend en entrée tous types d'objet (liste, dico, tuple, etc...) :
 * une liste : si int par ordre croissant, si str par ordre alphabétique (ordre ASCII (+ ou - alphabetique))
 * un tuple : si int par ordre croissant, si str par ordre alphabétique (ordre ASCII (+ ou - alphabetique))
 * un dictionnaire : via ses clés (si int par ordre croissant, si str par ordre alphabétique)
- Plus lent et moins performent car crée un nouvele liste
'''

sport = "bastien"
print("Trie un str :", sorted(sport))
sport1 =[2, 3, 1, 7, 6]
print("Trie une liste de int :", sorted(sport1))
print("Trie une liste de int à l'inverse :", sorted(sport1, reverse=True))
sport2 = {"accrobatique":"snow", "endurance":"trail", "collectif":"foot"}
print("Trie un dico par ses clés :", sorted(sport2))


  # 2) .sort(): que pour des liste!!!. Méthode ne pas confondre avec la fonction sorted()
'''
- retourne une nouvelle liste triée et laisse l'objet original inchangé.
- si vous avez besoin de conserver la liste d'origine intacte et que vous souhaitez obtenir une nouvelle liste triée.
- si on prend en entrée uniquement une liste :
 * si int par ordre croissant, si str par ordre alphabétique (ordre ASCII (+ ou - alphabetique))

'''
print("______ .sort() ______")
sport = list("bastien")
sport.sort()
print("Trie un str :", sport)
sport1 =[2, 3, 1, 7, 6]
sport1.sort()
print("Trie une liste de int :", sport1)
sport1.sort(reverse=True)
print("Trie une liste de int à l'inverse :", sport1)
sport2 = list({"accrobatique":"snow", "endurance":"trail", "collectif":"foot"})
sport2.sort()
print("Trie un dico par ses clés :", sport2)


# reverse() : fonction qui reverse
print("______ reversed() ______")
print(list(reversed(liste)))


# eval() : évalue un string comme une expression Python et retourne le résultat.
print("______ eval() ______")
print(eval("2+2"))
```

#**II- Les méthodes les plus vues pour chaine de caractères**

```python
liste1 = "Bonjour Le Jour"
liste2 ="Snow, Trail, Foot"
liste3 = "bonjour Le Jour"
liste7 = ["Anais", "Aubin", "Bastien", "Maman"]


# Revenir à  la ligne : \n


print("______ .lower(), .upper() et .capitalize()______")
#.lower : met toute la chaine en minuscule
print(".lower :", liste1.lower())
#.upper : met toute la chaine en majuscule
print(".upper :", liste1.upper())
#.capitalize() : met une majuscule au 1er mot
print(".capitalize :", liste3.capitalize())


print("______ .split()______")
#.split : split à "," en créant une liste si rien, split par défaut sur les espaces.
print("Ici rien en argument donc split sur les espace : ", liste2.split())
print("Ici, split sur ',' : ", liste2.split(","))
print("Ici, split la 2eme virugle sur 2eme mot : ", liste2.split(",")[1])


print("______ .replace()______")
#.replace : remplace dans une chaine de cractère mais pas une liste:
print(liste1.replace("jour", "soir")) #Remplace "jour" par "soir".
print(liste1.replace("jour", "soir").replace("Bon", "mauvais"))
print(liste2.replace("Snow","Ski")) #Remplace dans la liste "Snow" par "Ski".
print(liste2.replace("Foot","Ski").replace(",","/")) #Remplace à la chaine.


print("______ .find()______")
#.find : trouve l'index de la 1ere occurence d'un mot ou d'une lettre dans une chaine de str. !! trouve l'occurance en allant de gauche à droite.
  #Si "-1" c'est égal à "je ne trouve pas". Il prend en compte majuscule et minuscule
liste = "Bonjour Le Jour"
print("La 1ere lettre 'o' est à la place :", liste.find("o"))
print("-1 car ne trouve pas la 'g' car minuscule :", liste.find("g"))
  #On peut chercher des str pour valider un mdp par exemple
if liste.find("jo") > 0: #Pourquoi ""> 0" car si "< 0" soit "-1" ça n'existe pas !
  print("mdp juste")
else :
  print("mdp faux")


print("______ .strip()______")
#.strip (ou "enlever" en fr) : enlève les espaces à gauche et à droite de la chaine de caractère. Si on spécifie l'argument, enlève le caratère spécifé. Il exite .lstrip et .rstrip pour left et right:
print("      Bonjour le jour , strip()  ".strip())
print("---Bonjour le jour--, strip('-')---".strip("-"))
print("---Bonjour le jour--, lstrip('-')".lstrip("-"))
print("---Bonjour le jour--, rstrip('-')".rstrip("-"))


print("______ .isdigit()______")
#.isdigit : "is" avec "digit" sert à savoir si la chaine de caractère ne contient que des chiffres :
liste4 = "5070" # Que des chiffres donc ok
liste5 = "50 70" # Un espace donc pas que de chiffre donc faux
liste6 = "5070a" # Une lettre "a" donc pas que de chiffre donc faux
print("ici que des chiffres :", liste4.isdigit())
print("ici, un espace avec des chiffres :", liste5.isdigit())
print("ici, une lettre avec des chiffres :", liste6.isdigit())


print("______ .count()______")
#.count : Renvoie le nbre d'occurence de l'élément sépcifier dans un str, un int ou une liste attention c'est les caractères et non les mots:
liste = [1, 1, 2, "trois", 4, "cinq", 6]
print("L'occurence du chiffre '1' ds cette liste est de :", liste.count(1))
  #Autre exemple :
liste3 = "Bonjour le jour"
print("L'occurence de 'jour' ds ce str est de :", liste3.count("jour")) #2 occurence car pas d'espace et donc 2 fois "jour"
print("L'occurence de 'jour' ds ce str est de :", liste3.count(" jour ")) #1 occurence car des espaces et donc un seul " jour "


print("______ .startswith()______")
#.startswith (commence par..) : vérifier qu'un "str" commence par un ou plusieurs caractères.
url = "https://www.google.com".startswith("https")
url2 = "https://www.google.com".startswith(("https", "http", "www")) # On peut utiliser des "tuple" pour plusieurs vérifications.
print(url ,url2)


print("______ .join()______")
#.join : permet de convertir une liste en une chaîne de caractères, en utilisant un séparateur qu'on définit
'''La méthode .join() prend chaque élément de la liste et les concatène avec ", " en une seule chaîne de caractères.
'''
print(", ".join(liste7))
print("".join(liste7))


print("______ .reverse()______")
#reverse() : fonction qui reverse en créant une liste des caractères
print(list(reversed(liste)))


print("______ .enumerate()______")
# "enumerate" : récupère dans une liste de "tuple" le numéro de l'itération avec son élément associé à l'itération.
print(list(enumerate(liste7)))
print(list(enumerate(liste7, start=2))) # Start = décide de ou commence la numérotation

  #Exo : enumerate
'''
Prend une séquence (comme une liste) en entrée et renvoie une paire (index, valeur) pour chaque élément
de la séquence.
Du coup, on peut itérer à la fois sur l'index et sur la valeur de chaque élément de la liste.
'''
for i,y in enumerate(liste7, 1):
  test = f"Ligne n°{i}: {y}"
  print(test)
```

#**III- Les fonctions "enumerate()", "isenstance()"**

```python
# 1) enumarate():
'''énumérer les éléments d'une séquence (comme une liste, un tuple, ou une chaîne de caractères)
en leur attribuant un indice correspondant. Elle renvoie un objet énuméré qui peut être converti en
une liste de tuples contenant à la fois l'indice et l'élément correspondant de la séquence. enumerate(iterable, startIndex)'''

prenom2 = ["Anais", "Aubin", "Bastien"]

  # Exemple 1 : classique
liste = enumerate(prenom2, start=1) # enumere le 1er "str" soit "Anais"
print("Exemple 1 :", list(liste)) # Print un objet de type "enumarate" non printable donc on le passe en liste

  # Exemple 2: sans décalage de l'indice et avzec des listes
for i in enumerate(prenom2):
  print("Exemple 2 :", i) # Print en tuple par défaut sur de l'itération

  # Exemple 3 : avec un décalage de l'indice
for i, element in enumerate(prenom2, start=2):
    print(f"Exemple 3 : {i} est : {element}") # Print sans tuple ? Je sais pas pourquoi ?


# 2) isenstance():
'''Vérifie si un objet est une instance d'une "class" d'un type de données spécifié.
Elle prend deux arguments : l'objet que vous souhaitez tester et le type de données que vous voulez vérifier : isinstance(objet, type_de_donnees)'''

  # Vérifie si l'objet est une instance de int
print(isinstance(5, int))  # True
print(isinstance(5.5, int))  # False

  # Vérifie si l'objet est une instance de str
print(isinstance("hello", str))  # True
print(isinstance(5, str))  # False

  # Vérifie si l'objet est une instance de list
print(isinstance([1, 2, 3], list))  # True
print(isinstance((1, 2, 3), list))  # False

  # Vérifie si l'objet est une instance d'une classe utilisateur
class MaClasse:
    pass

obj = MaClasse()
print(isinstance(obj, MaClasse))  # True
print(isinstance(obj, int))  # False
```

#**IV- Les fonctions "lambda()", "map()" et "filter()"**

```python
# 1) lambda() :

  # a) Avec une fonction classique :
def times2(var):
    return var*2

times2(2)

  # b) Avec lambda :
lambda var: var*2


# 2) map() : map(fonction, l'itérable)
'''
Applique une fonction à chaque élément d'une séquence itérable (comme une liste ou une série).
'''
seq = [1,2,3,4,5]

map(times2,seq) # Renvoie un objet de la fonction sur seq

list(map(times2,seq)) # On transforme en liste

list(map(lambda var: var*2,seq)) # On remplace la fonction par "lambda"


# 3) filter() :

filter(lambda item: item%2 == 0,seq) # Idem, directement avec "lambda" sans passer par une fonction au préalable

list(filter(lambda item: item%2 == 0,seq)) # Idem, on transforme en liste
```
