#**II- La liste de course**

```python
import sys # Permet de sortir du script avec la methode ".exit" ds le choix 5

menu = '''Choisissez parmis les 5 options suivantes :
1: A jouter un √©l√®ment √† une liste
2: Retirer un √©l√©ment √† une liste
3: Afficher la liste
4: Vider la liste
5: Quitter
üëâ Votre choix : '''

l = []
choix_menu = ["1", "2", "3", "4", "5"] #L'entr√©e utilisateur (input) est toujours "str", donc pour comparer il faut que ce soir un liste en "str".

while True:
  demande = input(menu)

  while demande not in choix_menu :
    print("Le choix n'est pas valide")
    demande = input(menu)

  if demande == "1" :
    ajout = input("Quel aliment voulez-vous ajouter :").lower()  #.lower() permet de tout mettre en minuscule pour comparer en minucule et pas se tromper
    l.append(ajout)

  elif demande == "2" :
    item = input("Quel √©l√©ment voulez-vous retirer :").lower()
    if item in l :
      l.remove(item)
    else :
      print("L'√©l√©ment n'est pas dans liste")

  elif demande == "3" :
    print("Voici la liste")
    print(l)

  elif demande == "4" :
    print("Liste effac√©e")
    l.clear()

  elif demande == "5" :
    print("A bient√¥t !")
    sys.exit() #Quite le programme

  print("-" * 50) #Sert de s√©parateur afin (on imprime 50 x "-")
```
