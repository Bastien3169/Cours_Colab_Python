```python
# Les différents widgets
'''
CTkLabel : pour afficher du texte.
CTkButton : pour les boutons interactifs.
CTkEntry : pour les entrées de texte sur une seule ligne.
CTkSlider : pour les barres de défilement (sliders).
CTkCheckBox et CTkRadioButton : pour les cases à cocher ou les boutons radio.
CTkFrame : (cadre)
'''

# Padding, margin, height et width
'''
- height=40, width=40 : définit la taille interne du bouton, soit la hauteur et la largeur du bouton (assimilé au padding)
- padx=10, pady=10 : définit l'espacement externe autour du bouton, espace entre le bouton et les autres éléments (margin ou marges exterieures).
Le "padding" est l'espace entre le contenu (le texte dans le bouton) et les bords du bouton, mais dans ce cas,
height et width définissent directement la taille du bouton, pas le padding.
'''

# sticky et anchor
'''
anchor : aligne le texte à l'intérieur d'un widget.
sticky : aligne le widget dans la grille.

sticky="ew" : centré horizontalement
sticky="ns" : centré verticalement
sticky="nsew" : centré horizontalement et verticalement
sticky="e", "w", "n", "s" : centré à l'est, ouest, nord et sud
'''

# Méthodes de positionnement des widgets
'''
Méthode utilisée pour positionner un widget dans la fenêtre principale ou dans un conteneur (ou  gestionnaires de géométrie)
.pack() pour des interfaces simples ou pour empiler des widgets rapidement.
.grid() pour des interfaces structurées en lignes et colonnes.
.place() lorsque vous avez besoin d'un contrôle absolu sur la position des widgets (mais évitez-le si possible).
'''

# Gerer les variables des widgets avec .configure() et .get()/.set()
'''
.configure() : permet de modifier les propriétés ou attributs d'un widget après sa création.

.get() et .set()
Principalement utilisées avec des variables associées aux widgets (Entry, Label, Checkbutton, etc...)
(comme StringVar, IntVar, DoubleVar, ou BooleanVar) pour récupérer (.get) ou mettre à jour leur valeur (.set).

    EXEMPLE :
    - Création d'une variable Tkinter
    texte_variable = StringVar()
    texte_variable.set("Bonjour !")  # Initialisation de la valeur

    - Widget Label lié à la variable
    label = Label(app, textvariable=texte_variable)
    label.pack()

    - Récupérer la valeur actuelle
    print(texte_variable.get())  # Affiche : Bonjour !
'''
print("Cours")
```

#**II- CALCULETTES avec .configure() et .grid()**

```python
import customtkinter as ctk
import tkinter as tk

# Variable pour contenir l'expression mathématique
equation = ""

# Fonction pour gérer l'appui sur les boutons
def touche_appui(valeur):
    global equation

    if valeur == "=":
        try:
            # Évaluer l'expression et afficher le résultat
            resultat = str(eval(equation))
            equation = resultat
        except Exception:
            equation = "Erreur"
    else:
        # Ajouter la valeur du bouton à l'expression
        equation += str(valeur)

    # Mettre à jour l'écran
    ecran.configure(text=equation)

# Fonction pour supprimer la dernière lettre
def supprimer_derniere_lettre():
    global equation
    if len(equation) > 1:  # Si plus d'un caractère dans l'équation
        equation = equation[:-1]  # Supprime le dernier caractère
        # Mettre à jour l'écran
        ecran.configure(text=equation)
    else:
        equation = ""  # Réinitialise à "0" si tout est supprimé
        # Mettre à jour l'écran
        ecran.configure(text=equation)



# Créer la fenêtre principale avec CustomTkinter en appelant l'objet .Ctk()
app = ctk.CTk()

# Paramètres de la fenêtre (facultatif)
app.geometry("400x400")
app.minsize(400, 400)
app.title("Calculette Tkinter")
app.configure(fg_color = "pink", columnspan = 4) # avec tkinter c'est "background" et non "fg_color" pour foreground. "bg_color" pour background. Il y a aussi "hover_color"
app.grid_columnconfigure(0, weight=1)

# Afin d'éviter de configurer chaque colonne pour leur poid , on fait une boucle
for i in range(4): # Colonnes 0 à 3
    app.grid_columnconfigure(i, weight=1)


# Ecran
ecran = ctk.CTkLabel(app, text="0", font=('Arial', 40, 'bold'), text_color = "black", anchor="e", bg_color = "green")
ecran.grid(columnspan =4, sticky="ew") # Prend 4 colonnes. La 1ere est codée pour que ce soit la colonne "O"

# Les boutons
boutons = [7, 8, 9, "*", 4, 5, 6, "-", 1, 2, 3, "+", 0, ".", "/", "="]

ligne = 1 # 0 étant la ligne de l'écran
colonne = 0 # 0 n'existe pas car j'ai décalé avec le "range(1,4)" pour plus de lisibilité avec .grid()

for i in boutons :
    bout = ctk.CTkButton(app, text=str(i), font=('Arial', 24, 'bold'), text_color = "black", anchor="center", fg_color = "red", height=50, width=75, command=lambda val=i: touche_appui(val)) # Par défaut le boutton est bleu
    bout.grid(columnspan = 1, rowspan = 1, padx=10, pady=10, column=colonne, row=ligne, sticky="ew")

    colonne = colonne + 1 # (ou colonne += 1)
    if colonne > 3 :
        colonne = 0
        ligne = ligne + 1

# Le bouton effacer
dele = ctk.CTkButton(app, text="DEL", font=('Arial', 24, 'bold'), text_color = "black", anchor="center", fg_color = "yellow", height=50, width=75, command= supprimer_derniere_lettre) # Par défaut le boutton est bleu
dele.grid(columnspan = 4, rowspan = 1, padx=10, pady=10, column=colonne, row=ligne, sticky="ew")

# Boucle principale pour afficher la fenêtre
app.mainloop()
```

#**IV- Exemples : .get()/.set() et .configure()**

```python
# Fenêtre principale
app = Tk()


# Fonction pour mettre à jour la valeur
def changer_texte():
    label.configure(text="Merci d'avoir cliqué !")  # Met à jour directement le texte du Label

# Widget Label lié à la variable
label = Label(app, text="Cliquez sur le bouton !")
label.pack()

# Bouton pour déclencher le changement
button = Button(app, text="Changer le texte", command=changer_texte)  # PAS de parenthèses ici !
button.pack()

app.mainloop()
```

```python
# Fenêtre principale
app = Tk()

# Fonction pour mettre à jour la valeur
def changer_texte():
    texte_variable.set("Merci d'avoir cliqué !")

# Création d'une variable StringVar()
texte_variable = StringVar()
texte_variable.set("Changer le texte")

# Widget Label lié à la variable
label = Label(app, text="Cliquez sur le bouton !")  # Lien avec la variable
label.pack()

# Bouton pour déclencher le changement
button = Button(app, textvariable=texte_variable, command=changer_texte)  # PAS de parenthèses ici !
button.pack()

app.mainloop()
```
